# 元组

**学习目标：** 本节介绍元组的创建和常用操作，重点比较元组与列表在可变性和使用场景上的差异。


## 元组基础


在 `Python` 中，元组 `tuple` 与列表相似，但元组是不可变对象，创建后不能修改其中的元素。

元组适合表示不应被修改的数据，例如图像尺寸、日期片段、坐标点或函数返回的多项结果。

本节主要讨论：

1. 元组创建方法。
2. 元组常用操作。
3. 元组与列表的差异。


## 元组创建


元组使用圆括号 `()` 表示，元素之间用逗号分隔。


```python
tuple_1 = (1, 2, 3)
tuple_2 = ('python',100)
tuple_3 = 'a','b','c',1, 2, 3
print(tuple_1)
print(tuple_2)
print(tuple_3)
```

当然，也可以创建空元组，如：


```python
empty_tuple = tuple()
print(empty_tuple)
```

但是，当创建的元组只有一个元素时，需要加上逗号，例如：


```python
single_tuple = (1, )
print(single_tuple)
print(type((1)))
print(type((1, )))
```

`len()` 函数可以获取元组有多少项。


```python
my_tuple = (1, 2, 3)
print(len(my_tuple))
```

## 元组操作


元组操作也有很多内置函数，当然，没有列表那么多操作方法。下面介绍几个常见的方法。


- 访问元组
元组可以使用下标索引来访问元组中的值，如：


```python
tuple_1 = ('Python', 'C++', 60, 59)
tuple_2 = (1, 2, 3, 4, 5, 6, 7)
print(tuple_1[0])
print(tuple_2[3:6])
```

- 修改元组
元组中的元素值是不允许修改的，但我们可以对元组进行连接组合，如:


```python
tuple_1 = (12, 34.56)
tuple_2 = ('abc', 'xyz')

# 以下修改元组元素操作是非法的。
# tuple_1[0] = 100

# 创建一个新的元组
tuple_3 = tuple_1 + tuple_2
print(tuple_3)
```

- 删除元组
元组中的元素值是不允许删除的，但我们可以使用 del 语句来删除整个元组，如:


```python
tuple_1 = ('Python', 'C++', 60, 59)
print(tuple_1)
del tuple_1
print(tuple_1)
```

- 元素索引
通过 index 方法可以获取元素在元组中的索引，如：


```python
tuple_1 = ('Python', 'C++', 60, 59)
ind = tuple_1.index('C++')
print(ind)
```

- 元素个数统计
通过 count 方法可以统计元组中元素出现的次数，如：


```python
tuple_1 = (1, 2, 3, 4, 4, 4, 3, 3, 2, 1)
cnt = tuple_1.count(4)
print(cnt)
```

## 小结


在实际开发中，元组不像列表那样经常使用，但是在需要不可变性时才会使用。如果在程序中传递一个对象并需要确保它不会被更改，使用 tuple 会比较方便。

### 变量命名

`Python` 命名需要遵循以下规则：

- 变量名只能包含字母、数字和下划线 `_`。
- 变量名可以用字母或下划线 `_` 开头，但不能以数字开头。
- 变量名不能包含空格，可以用下划线 `_` 分隔单词。
- 变量名不能使用 `Python` 关键字，也不应该覆盖内置函数名。
- 变量名应具有描述性，含义明确，同时保持简洁。
- 慎用小写字母 `l` 和大写字母 `O`，避免与数字 `1` 和 `0` 混淆。

### 变量、对象与类型

在 `Python` 中，变量本身没有固定类型，变量只是绑定到对象的名称；真正拥有类型的是对象。


在 `C++` 等语言中，变量通常有固定类型。例如：

```cpp
int a = 3;
```

此时变量 `a` 只能存储 `int` 类型的数据，下面这种赋值在编译阶段就会报错：

```cpp
a = "hello";
```

但在 `Python` 中，变量没有固定类型。使用变量前不需要提前声明类型，直接把对象赋值给变量即可：

```python
a = 3
a = "hello"
```

需要注意，不能直接使用没有赋值的变量。由于变量本身没有固定类型，同一个变量名可以在不同时间绑定到不同类型的对象。

下面的语句看起来像是在查看变量 `a` 的类型：

```python
a = 3
print(type(a))
```

实际上，`type(a)` 返回的是变量 `a` 当前引用对象的类型，也就是整数对象 `3` 的类型。


### 引用

在 `Python` 中，变量保存的是对象引用。可以把变量理解为指向对象的名称，而不是直接存储对象本身。

```python
x = [1, 2, 3]
y = x

print(x is y)  # True，说明 x 和 y 引用同一个列表对象。
```


对象是内存中存储数据的实体，变量是绑定到对象的名称。当执行 `a = 3` 时，`Python` 会创建整数对象 `3`，然后让变量 `a` 引用这个对象。

因此，执行 `print(type(a))` 时，输出的是对象 `3` 的类型，而不是变量名 `a` 自身的类型。


### 可变对象与不可变对象


````python
a = 3
print( hex(id(a)))
a = 5
print( hex(id(a)))
````

`id()` 返回对象的内存标识，`hex()` 可以把整数转换为十六进制字符串。执行 `a = 3` 时，变量 `a` 引用整数对象 `3`；再执行 `a = 5` 时，变量 `a` 改为引用另一个整数对象 `5`。因此，两次输出的对象标识可能不同。


````python
a = {'name': 'c++'}
print( hex(id(a)))
a['name'] = 'python'
print( hex(id(a)))
````

变量 `a` 最初引用一个字典对象，其中键是 `"name"`，值是 `"c++"`。当执行 `a["name"] = "python"` 时，字典内容发生变化，但字典对象本身没有被替换，所以 `a` 仍然引用同一个对象。


## 运算符

### 算术运算符

| 符号 | 含义 |
| :--- | :--- |
| `+` | 两个对象相加。 |
| `-` | 得到负数，或计算两个数的差。 |
| `*` | 两个数相乘，或返回重复若干次的字符串。 |
| `/` | 除法，返回浮点数结果。 |
| `%` | 返回除法的余数。 |
| `**` | 返回 `x` 的 `y` 次幂。 |
| `//` | 整除，返回商的整数部分。 |


````python
a = 24
b = 5
c = 0

print(f"a = {a}, b = {b}")

c = a + b
print ("1. a + b = ", c)

c = a - b
print ("2. a - b = ", c)

c = a * b
print ("3. a * b = ", c)

c = a / b
print ("4. a / b = ", c)

c = a // b
print ("4.1 a // b = ", c)

c = a % b
print ("5. a % b = ", c)

# 更新变量
a = 2
b = 3
c = a**b
print ("6. a ** b = ", c)

a = 10
b = 5
c = a//b
print ("7. a // b = ", c)
````

> **注意**
>
> 1. `Python` 可以同时为多个变量赋值，例如 `a, b = 1, 2`。
> 2. 一个变量可以通过赋值绑定到不同类型的对象。
> 3. 数值除法包含两个运算符：`/` 返回浮点数，`//` 返回向下取整后的整数。
> 4. 取余可以表示为 `r = a - n * (a // n)`，其中 `r` 是余数，`a` 是被除数，`n` 是除数。
> 5. 混合计算时，`Python` 通常会把整数转换为浮点数。


````python
print(5//2)
print(-5//2)
print(123%10)
print(-123%10)
print(123%-10)
````

### 比较运算符

| 符号 | 含义 |
| :--- | :--- |
| `==` | 比较两个对象是否相等。 |
| `!=` | 比较两个对象是否不相等。`Python 3.x` 已取消 `<>` 运算符。 |
| `>` | 判断左侧值是否大于右侧值。 |
| `<` | 判断左侧值是否小于右侧值。 |
| `>=` | 判断左侧值是否大于等于右侧值。 |
| `<=` | 判断左侧值是否小于等于右侧值。 |

所有比较运算符都会返回布尔值 `True` 或 `False`。


````python
a = 21
b = 10
print(f"a = {a}, b = {b}")

if ( a == b ):
    print ("1. a 等于 b")
else:
    print ("1. a 不等于 b")

if ( a != b ):
    print ("2. a 不等于 b")
else:
    print ("2. a 等于 b")

if ( a < b ):
    print ("3. a 小于 b")
else:
    print ("3. a 大于等于 b")

if ( a > b ):
    print ("4. a 大于 b")
else:
    print ("4. a 小于等于 b")

# 更新变量
a = 5
b = 20
print(f"a is {a}, b is {b}")
if ( a <= b ):
    print ("5. a 小于等于 b")
else:
    print ("5. a 大于  b")

if ( b >= a ):
    print ("6. b 大于等于 a")
else:
    print ("6. b 小于 a")
````

### 位运算符

按位运算符会把数字按二进制位进行计算。假设 `a = 60`、`b = 13`：

- `&`：按位与。对应位都为 `1` 时，结果位为 `1`，例如 `a & b` 的结果是 `12`。
- `|`：按位或。对应位至少有一个为 `1` 时，结果位为 `1`，例如 `a | b` 的结果是 `61`。
- `^`：按位异或。对应位不同时，结果位为 `1`，例如 `a ^ b` 的结果是 `49`。
- `~`：按位取反。`~x` 等价于 `-x - 1`，例如 `~a` 的结果是 `-61`。
- `<<`：左移运算符，例如 `a << 2` 的结果是 `240`。
- `>>`：右移运算符，例如 `a >> 2` 的结果是 `15`。


````python
a = 60  # 60 = 0011 1100
b = 13  # 13 = 0000 1101
c = 0

c = a & b  # 12 = 0000 1100
print ("1. c 的值为：", c)

c = a | b  # 61 = 0011 1101
print ("2. c 的值为：", c)

c = a ^ b  # 49 = 0011 0001
print ("3. c 的值为：", c)

c = ~a  # -61 = 1100 0011
print ("4. c 的值为：", c)

c = a << 2  # 240 = 1111 0000
print ("5. c 的值为：", c)

c = a >> 2  # 15 = 0000 1111
print ("6. c 的值为：", c)
````

### 逻辑运算符

假设 `a = 10`、`b = 20`：

- `and`：如果 `x` 为 `False`，`x and y` 返回 `False`；否则返回 `y` 的计算值。例如 `a and b` 返回 `20`。
- `or`：如果 `x` 为 `True`，`x or y` 返回 `x` 的值；否则返回 `y` 的计算值。例如 `a or b` 返回 `10`。
- `not`：如果 `x` 为 `True`，`not x` 返回 `False`；如果 `x` 为 `False`，`not x` 返回 `True`。例如 `not (a and b)` 返回 `False`。


````python
a = 10
b = 20

print(f"a = {a}, b = {b}")
if  a and b :
    print ("1. 变量 a 和 b 都为 True")
else:
    print ("1. 变量 a 和 b 有一个不为 True")

if ( a or b ):
    print ("2. 变量 a 和 b 都为 True，或其中一个变量为 True")
else:
    print ("2. 变量 a 和 b 都不为 true")

# 更新变量
a = 0
print(f"a = {a}, b = {b}")

if ( a and b ):
    print ("3. 变量 a 和 b 都为 True")
else:
    print ("3. 变量 a 和 b 有一个不为 True")

if ( a or b):
    print ("4. 变量 a 和 b 都为 True，或其中一个变量为 True")
else:
    print ("4. 变量 a 和 b 都不为 true")

if not( a and b ):
    print ("5. 变量 a 和 b 都为 false，或其中一个变量为 false")
else:
    print ("5. 变量 a 和 b 都为 true")
````

### 成员运算符

- `in`：如果在指定序列中找到值，则返回 `True`；否则返回 `False`。
- `not in`：如果在指定序列中没有找到值，则返回 `True`；否则返回 `False`。


````python
a = 10
b = 20
print(f"a is {a}, b is {b}")
mylist = [1, 2, 3, 4, 5]

if ( a in mylist ):
   print ("1. 变量 a 在给定的列表中 list 中")
else:
   print ("1. 变量 a 不在给定的列表中 list 中")

if ( b not in mylist ):
   print ("2. 变量 b 不在给定的列表中 list 中")
else:
   print ("2. 变量 b 在给定的列表中 list 中")

# 更新变量
a = 2
print(f"a is {a}, b is {b}")

if ( a in mylist ):
   print ("3. 变量 a 在给定的列表中 list 中")
else:
   print ("3. 变量 a 不在给定的列表中 list 中")
````

### 身份运算符

- `is`：判断两个标识符是否引用同一个对象。`x is y` 类似于 `id(x) == id(y)`。
- `is not`：判断两个标识符是否引用不同对象。`x is not y` 类似于 `id(x) != id(y)`。


> **注意**
>
> - 对于较小整数，`Python` 可能复用同一个对象；因此值相等时，`is` 也可能返回 `True`。
> - 不要用 `is` 比较普通数值是否相等。比较值是否相等应使用 `==`，只有判断是否为同一个对象时才使用 `is`。


````python
# == 比较值，is 比较对象身份
a = [1, 2, 3]
b = a
c = [1, 2, 3]

print(a == c)
print(a is c)
print(a is b)
print(id(a), id(b), id(c))
````

### 运算符优先级

从高到低，常见运算符优先级如下：

| 运算符 | 含义 |
| :--- | :--- |
| `**` | 指数运算。 |
| `~`、`+`、`-` | 按位取反、一元加号、一元减号。 |
| `*`、`/`、`%`、`//` | 乘、除、取模、取整除。 |
| `+`、`-` | 加法、减法。 |
| `>>`、`<<` | 右移、左移。 |
| `&` | 按位与。 |
| `^`、`|` | 按位异或、按位或。 |
| `<=`、`<`、`>`、`>=` | 比较运算符。 |
| `==`、`!=` | 相等性比较。 |
| `=`、`%=`、`/=`、`//=`、`-=`、`+=`、`*=`、`**=` | 赋值运算符。 |
| `is`、`is not` | 身份运算符。 |
| `in`、`not in` | 成员运算符。 |
| `and`、`or`、`not` | 逻辑运算符。 |


````python
a = 20
b = 10
c = 15
d = 5
e = 0

e = (a + b) * c / d  # ( 30 * 15 ) / 5
print ("(a + b) * c / d 运算结果为：",  e)

e = ((a + b) * c) / d  # (30 * 15 ) / 5
print ("((a + b) * c) / d 运算结果为：",  e)

e = (a + b) * (c / d);  # (30) * (15/5)
print ("(a + b) * (c / d) 运算结果为：",  e)

e = a + (b * c) / d;  #  20 + (150/5)
print ("a + (b * c) / d 运算结果为：",  e)
````

## 字符串

字符串是 `Python` 中最常用的数据类型之一。可以使用单引号 `'`、双引号 `"` 或三引号 `"""` 创建字符串。

创建字符串很简单，只要为变量分配一个字符串值即可。例如：


````python
str_1 = 'hello, python'
str_2 = "hello, python"
str_3 = """hello, python"""
````

可以打印字符串，也可以使用 `type()` 查看变量当前引用对象的类型。


````python
print(str_1, type(str_1))
print(str_2, type(str_2))
print(str_3, type(str_3))
````

### 索引的概念


`Python` 中，字符串是字符的有序集合。字符串中的每个字符都有自己的位置，也就是索引。

例如，字符串 `"python"` 中第 3 个字符是 `"t"`。需要注意，`Python` 的索引从 `0` 开始，因此 `"t"` 的索引是 `2`。


### 字符串的切片操作


切片是常用的序列操作，基本形式是 `sequence[start:end:step]`。

- `start` 表示开始索引，默认值为 `0`。
- `end` 表示结束索引，默认值为序列长度。
- `step` 表示步长，默认值为 `1`。

切片遵循左闭右开原则，即包含 `start` 位置，不包含 `end` 位置。


- 指定开始和结束位置


````python
a = "hello crash course"
print(a[6:11])
print(a[6:-7])
````

- 指定开始位置，不指定结束位置


````python
a = "hello crash course"
print(a[-1:5:-1])
````

- 指定结束位置，不指定开始位置


````python
a = "hello crash course"
print(a[:-7])
````

- **切片操作允许索引越界**


````python
a = "hello crash course"
print(len(a))
print(a[7:20])
````

字符串 `a` 的长度是 `18`，索引从 `0` 开始。切片允许索引越界；当结束索引超过字符串长度时，`Python` 会自动截取到字符串末尾。


- 指定切片间隔


````python
a = '123456789'
print(a[::2])
print(a[1::2])
print(a[::-2])
print(a[::-1])
````

`[]` 操作符中可以出现两个冒号 `:`。第一个冒号用于分隔开始索引和结束索引，第二个冒号用于分隔结束索引和步长。

当开始索引和结束索引都省略时，等价于从字符串开头截取到字符串末尾。


综合应用例子


````python
str_1 = 'Ronaldo'
print(len(str_1))
print(str_1)
print(str_1[-1])
# -1 等价于 len(str) - 1
print(str_1[0:-1])  # 输出第一个到倒数第二个的所有字符
print(str_1[0]) # 输出第一个字符
print(str_1[2:5]) # 输出从第三个开始到第五个的字符
print(str_1[2:]) # 输出从第三个开始后的所有字符
print((str_1+str_1[-1])*10)
print(str_1 + 'TEST') # 链接字符串
print('g' + 'o'*100 + 'al')
print(str_1[6:-1])
print(str_1[2:])
mystr = 'aa  b  c  d'
print(mystr.split('  '))
````

### 字符串与基本类型转换

````python
a = 100
print(str(a)) # int 转 str
b = 100.1
print(str(b)) # float 转 str
c = True
print(str(c)) # bool 转 str
print(bool('True')) # str 转 bool
print(bool('False')) # str 转 bool
print(bool(' '))  # str 转 bool
print(bool(''))  # 非空字符串为 True
print(int('123'))  # str 转 int
print(float('123'))  # str 转 float
print(float('123.4')) # str 转 float
print(int('123.4'))  # 小数字符串转换为整数会报错
````

## 进制转换


我们平时使用的是十进制，规则是逢 `10` 进 `1`。计算机底层常用二进制；除此之外，程序中也经常遇到八进制和十六进制。

### 手动计算进制转换


````python
print(0b1101010)  # 二进制
print(0o125)  # 八进制
print(0x3f2a1)  # 十六进制
````

### 使用函数进行进制转换


````python
value = int('1101010', 2)
print(value)  # 106

value = int('125', 8)
print(value)  # 85

value = int('3f2a1', 16)
print(value)  # 258721
````

````python
print(bin(106))  # 转二进制
print(oct(85))  # 转八进制
print(hex(258721))  # 转十六进制
````

````python
print(bin(int('125', 8)))  # 八进制转二进制
print(bin(int('0x3f2a1', 16)))  # 十六进制转二进制
````

````python
print(oct(0b1101010))  # 二进制转八进制
print(hex(int('1101010', 2)))  # 二进制转十六进制
````

````python
print(oct(0x3f2a1))  # 十六进制转八进制
print(hex(int('125', 8)))  # 八进制转十六进制
````

## 文件操作

### 文件类型


`Python` 可以处理两类文件：文本文件和二进制文件。文本文件以字符编码保存内容，适合逐行读取和编辑；二进制文件以字节保存内容，适合图片、文档、数据库等非纯文本数据。

常见二进制文件包括：

- 图像文件：`.jpg`、`.png`、`.bmp`、`.gif`。
- 数据库文件：`.mdb`、`.frm`、`.sqlite`。
- 文档文件：`.doc`、`.xls`、`.pdf`。


不同文件类型需要不同的处理方式。例如，`.xls` 文件通常需要表格软件或相关库处理，`.sqlite` 文件需要数据库工具处理。

文本文件通常可以用标准文本编辑器打开。一个规范的文本文件通常满足以下特点：

- 内容可读，即使包含 `HTML` 或其他标记语言，也能以文本形式查看。
- 数据按行组织，大多数情况下每一行都是一个独立元素。
- 行尾包含换行符，在 `Python` 字符串中通常表示为 `"\n"`。


### 打开文件


在 `Python 3` 中，`open()` 会返回一个文件对象。基础用法如下：

```python
file_object = open(filename, mode)
```


`filename` 是要操作的文件名，通常包含扩展名。例如，文件名应写成 `"workData.txt"`，而不是只写 `"workData"`。

也可以传入完整路径。`Windows` 路径中的反斜杠 `\` 在 `Python` 字符串中有转义含义，因此推荐使用以下写法之一：

```python
path = r"C:\ThisFolder\workData.txt"
path = "C:/ThisFolder/workData.txt"
```

现代代码中，也可以优先使用 `pathlib.Path` 构造跨平台路径。


`open()` 函数中的 `mode` 参数告诉 `Python` 要用哪种方式处理文件。常见文本模式如下：

- `w`：写入模式。会覆盖已有文件内容，文件指针位于开头。
- `r`：读取模式。只读取文件内容，文件指针位于开头。
- `a`：追加模式。写入内容会追加到文件末尾。
- `r+`：读写模式。可读取也可修改文件，文件指针位于开头。
- `a+`：追加和读取模式。可读取文件，也可在文件末尾追加内容。


在使用二进制文件时，将使用相同的模式说明符。但是，在末尾添加一个 b。所以二进制文件的写模式说明符是'wb'。其他的分别是'rb'，'ab'，'r+b'和'a+b'。


`Python 3` 提供了独占创建模式 `"x"`。该模式专门用于创建新文件；如果同名文件已经存在，`open()` 会抛出异常，避免意外覆盖。


在使用 `open()` 函数时，通常会将其结果分配给变量。给定一个名为 workData.txt 的文件，打开文件进行读写的正确代码如下:


````python
data_file = open("workData.txt", "r+")
````

这会创建一个名为 `data_file` 的文件对象，然后可以使用文件对象方法对其进行操作。

示例中使用的 `r+` 模式表示读写已有文件。实际项目中，如果只需要读取文件，应该使用 `"r"`；如果只需要写入新内容，应该使用 `"w"` 或 `"a"`，减少误操作风险。


````python
data_file.read()
````

如果再读一遍会发生什么？


````python
data_file.read()
````

可以发现再读一遍之后，内容是空的。这是因为在读取文件后，读取的“游标”位于文件的末尾。所以没有什么可读的了。
我们可以这样重置“光标”:


````python
data_file.seek(0)
````

````python
#重读一次
print(data_file.read())
````

打开文件一般需要用 close 关闭文件。


````python
data_file.close()
````

````python
data_file.read()
````

在 `Python` 中，打开和关闭文件的最佳实践是使用 `with` 语句。`with` 代码块结束后，文件会自动关闭。


````python
with open("workData.txt", "r+") as workData:
    # File object is now open.
    # Do stuff with the file:
    print(workData.read())

# File object is now closed.
# Do other things...
````

如果不使用 `with` 语句，也可以手动调用 `file_object.close()` 关闭文件。但手动关闭容易遗漏，尤其是在代码抛出异常时。

因此，推荐使用 `with open(...) as file_object:` 管理文件生命周期，让文件在代码块执行结束后自动关闭。


### 读取文件


````python
with open("workData.txt", "r+") as work_data:
    print("This is the file name: ", work_data.name)
    line = work_data.read()
    print(line)
````

### `readline()` 一行一行读


````python
with open("workData.txt", "r+") as work_data:
     print("This is the file name: ", work_data.name)
     line_data = work_data.readline()
     print(line_data)
     print(work_data.readline())
````

你可以不断重复上述过程。


类似的方法还有 `file_object.readlines()`。注意方法名是复数形式，它会一次性读取所有行，并返回由字符串组成的列表。


````python
with open("workData.txt", "r+") as work_data:
    print(work_data.readlines())
````

这将把整个文件读入内存，并将其分割为几行。但是这只适用于文本文件。二进制文件只是一团数据——它并没有真正理解一行是什么。


在 `Python` 中，逐行处理文本文件的常见方式是直接遍历文件对象：


````python
with open("workData.txt", "r+") as work_data:
    for line in work_data:
        print(line, end = '')
````

### 写入文件


如果不能向文件写入数据，文件就没有任何用处。所以我们来讨论一下如何写文件。


> 如果用会创建文件的模式打开文件，而文件不存在，`Python` 会创建该文件。首次创建文件时，常见模式包括 `"a+"` 和 `"w+"`。其中 `"w+"` 会清空已有文件内容，使用前需要确认不会覆盖重要数据。


通常最好使用 a+模式，因为数据将默认添加到文件的末尾。使用 w+将清除文件中的任何现有数据，并给你一个空白文件开始。


````python
with open("workData.txt", "a+") as work_data:
    work_data.write("This data is on line 4\n")
````

如果您想将一些不是字符串的内容写入文本文件，例如一系列数字，您必须使用转换代码将它们转换或“强制转换”为字符串。
例如，如果您想要将整数 1234、5678、9012 添加到 work_data 文件中，您需要执行以下操作。首先，你把你的非字符串转换为字符串，然后你写字符串到你的文件对象:


````python
values = [1234, 5678, 9012]

with open("workData1.txt", "a+") as work_data:
    for value in values:
        str_value = str(value)
        work_data.write(str_value)
        work_data.write("\n")
````

### 编辑文件


有时需要编辑已有文件，而不是只向文件末尾追加数据。不能简单依赖 `"w+"` 模式，因为 `"w"` 会先清空文件内容；`"a+"` 又总是在文件末尾写入。

如果要修改文件中间内容，常见做法是读取原内容，生成修改后的新内容，再整体写回文件。


最简单的方法是取出整个文件并使用它创建一个列表或数组数据类型。一旦创建了列表，就可以使用该列表了。`insert(i, x)` 方法插入新数据。一旦创建了新的列表，您就可以将其合并在一起，并将其写入您的文件。


例如，使用我们的 `workData.txt` 文件，假设我们需要在第一行和第二行之间插入文本“This goes between line 1 and 2”。代码是:


````python
# 读取原文件内容。
with open("workData.txt", "r") as work_data:
    lines = work_data.readlines()

lines.insert(1, "This goes between line 1 and 2\n")
updated_content = "".join(lines)

# 覆盖写回修改后的内容。
with open("workData.txt", "w") as work_data:
    work_data.write(updated_content)

with open("workData.txt", "r") as work_data:
    print(work_data.read())
````

## 目录操作


### 获取目录


假设你当前的工作目录有一个叫 my_directory 的子目录，该目录包含如下内容：


├── `file1.py`
├── `file2.csv`
├── `file3.txt`
├── sub_dir
│   ├── `bar.py`
│   └── `foo.py`
├── sub_dir_b
│   └── `file4.txt`
└── sub_dir_c
    ├── `config.py`
    └── `file5.txt`


`Python` 内置的 `os` 模块提供了许多列出目录内容和过滤结果的方法。要获取目录中的文件和文件夹列表，可以使用 `os.scandir()`；如果只需要名称列表，也可以使用 `os.listdir()`。

如果还需要文件大小、修改时间等属性，`os.scandir()` 通常比 `os.listdir()` 更合适。


````python
import os

[os.makedirs('my_directory\\sub_dir' + dir_name) if not os.path.exists('my_directory\\sub_dir' + dir_name) else None for dir_name in ['', '_b', '_c']]
[open('my_directory\\' + file_name, 'a+').close() if not os.path.exists('my_directory\\' + file_name) else None for file_name in ['file1.py', 'file2.csv', 'file3.txt', 'config.py', 'file5.txt', 'sub_dir\\bar.py', 'sub_dir\\foo.py', 'sub_dir_b\\file4.txt']]

entries = os.listdir('my_directory')
print(entries)
````

目录列表现在看上去不容易阅读，对 `os.listdir()` 的调用结果使用循环打印有助于查看。


````python
for entry in entries:
    print(entry)
````

在现代 `Python` 版本中，推荐优先使用 `pathlib.Path` 处理路径；如果需要较底层的目录扫描，可以使用 `os.scandir()`。


````python
import os
entries = os.scandir('my_directory')
print('\n'.join([f.name for f in entries]))
entries.close()
````

````python
with os.scandir('my_directory') as entries:
    print('\n'.join([entry.name for entry in entries]))
````

这里 `os.scandir()` 和 with 语句一起使用，因为它支持上下文管理协议。使用上下文管理器关闭迭代器并在迭代器耗尽后自动释放获取的资源。在 my_directory 打印文件名的结果就和在 `os.listdir()` 例子中看到的一样


另一个获取目录列表的方法是使用 pathlib 模块：


````python
from pathlib import Path

for entry in Path(r"C:\\Users\\Maxwell\\Dropbox\\Python\\TeachingMaterials").iterdir():
    print(entry.name)
````

`pathlib.Path()` 会根据操作系统返回 `PosixPath` 或 `WindowsPath` 对象。

`Path` 对象的 `iterdir()` 方法会返回一个迭代器，用于遍历该目录下的文件和子目录。每个条目仍然是 `Path` 对象，因此可以继续调用 `is_file()`、`is_dir()`、`stat()` 等方法。

`pathlib` 从 `Python 3.4` 开始引入，为文件系统路径提供了面向对象接口。相比手动拼接字符串或混用多个 `os.path` 函数，`pathlib` 通常更清晰，也更容易写出跨平台代码。


使用 `pathlib.Path()` 或 `os.scandir()` 来替代 `os.listdir()` 是获取目录列表的首选方法，尤其是当你需要获取文件类型和文件属性信息的时候。`pathlib.Path()` 提供了在 os 和 shutil 中大部分处理文件和路径的功能，并且它的方法比这些模块更加有效。我们将讨论如何快速的获取文件属性。


| 函数 | 描述 |

| `os.listdir()` | 以列表的方式返回目录中所有的文件和文件夹 |
| `os.scandir()` | 返回一个迭代器包含目录中所有的对象，对象包含文件属性信息 |

### 列出目录中的所有文件


这节将向你展示如何使用 `os.listdir()` ，`os.scandir()` 和 `pathlib.Path()` 打印出目录中文件的名称。为了过滤目录并仅列出 `os.listdir()` 生成的目录列表的文件，要使用 `os.path` 。


````python
import os

base_path = 'my_directory'
for entry in os.listdir(base_path):
    # 使用os.path.isfile判断该路径是否是文件类型
    if os.path.isfile(os.path.join(base_path, entry)):
        print(entry)
````

在这里调用 `os.listdir()` 返回指定路径中所有内容的列表，接着使用 `os.path.isfile()` 过滤列表让其只显示文件类型而非目录类型。


一个更简单的方式来列出一个目录中所有的文件是使用 `os.scandir()` 或 `pathlib.Path()`


````python
import os

base_path = 'my_directory'
with os.scandir(base_path) as entries:
    for entry in entries:
        if entry.is_file():
            print(entry.name)

print('\n', end = '')

with os.scandir(base_path) as entries:
    for entry in entries:
        if entry.is_dir():
            print(entry.name)
````

使用 `os.scandir()` 比起 `os.listdir()` 看上去更清楚和更容易理解。对 `ScandirIterator` 的每一项调用 `entry.isfile()` ，如果返回 `True` 则表示这一项是一个文件。


接着，展示如何使用 `pathlib.Path()` 列出一个目录中的文件：


````python
from pathlib import Path

base_path = Path('my_directory')
for entry in base_path.iterdir():
    if entry.is_file():
        print(entry.name)
````

如果将 `for` 循环和 `if` 语句组合成单个生成器表达式，则上述的代码可以更加简洁。


````python
from pathlib import Path

base_path = Path('my_directory')
files_in_basepath = (entry for entry in base_path.iterdir() if entry.is_file())
for item in files_in_basepath:
    print(item.name)
````

````python
print('\n'.join(item.name for item in Path('my_directory').iterdir() if item.is_file()))
````

### 列出子目录


如果要列出子目录而不是文件，请使用下面的方法。现在展示如何使用 `os.listdir()` 和 `os.path()` :


````python
import os

basepath = 'my_directory'
for entry in os.listdir(basepath):
    if os.path.isdir(os.path.join(basepath, entry)):
        print(entry)
````

当你多次调用 `os.path,join()` 时，以这种方式操作文件系统就会变得很笨重。下面是如何使用 `os.scandir()` ：


````python
import os

base_path = 'my_directory'
with os.scandir(base_path) as entries:
    for entry in entries:
        if entry.is_dir():
            print(entry.name)
````

与文件列表中的示例一样，此处在 `os.scandir()` 返回的每一项上调用 `is_dir()` 。如果这项是目录，则 `is_dir()` 返回 `True`，并打印出目录的名称。输出结果和上面相同。


下面是如何使用 `pathlib.Path()` ：


````python
from pathlib import Path

base_path = Path('my_directory')
for entry in base_path.iterdir():
    if entry.is_dir():
        print(entry.name)
````

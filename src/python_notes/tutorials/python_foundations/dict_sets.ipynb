## 字典


我们已经学习了 `Python` 中的序列 `sequence`。本节开始学习映射 `mapping`，其中最常用的类型是字典 `dict`。

字典是键值对集合，用于存储映射关系。你可以把它理解为其他语言中的哈希表：通过键 `key` 快速找到对应的值 `value`。


### 字典创建


字典的每个键值 key=>value 对用冒号':' 分割，每个键值对之间用逗号‘,’分割，整个字典包括在花括号 '{} '中，格式如下所示：


```python
my_dict = {'key1':'value1', 'key2':'value2'}
print(my_dict)
```

字典也可以通过内置函数 `dict()` 创建；空字典可以用 `{}` 创建。


```python
# 空字典
my_dict = {}
print("Empty Dictionary: ")
print(my_dict)

# 键值对创建
my_dict = dict([(1, 'a'), (2, 'b'), (3, 'c')])
print("\nDictionary with each item as a pair: ")
print(my_dict)

# dict() 创建
my_dict = dict({'key1':'value1','key2':'value2'})
print("\nDictionary with the use of dict(): ")
print(my_dict)
```

字典可以嵌套，例如：


```python
{'key1':'value1','key2': dict([(1, 'a'), (2, 'b'), (3, 'c')])}
```

> 字典也可以使用推导式创建。


```python
squares = {x: x*x for x in range(20) if x % 2 == 1}

print(squares)
```

创建字典后，可以通过键来访问值，例如：


```python
print(my_dict['key2'])
```

> 字典的键通常应保持唯一。如果写入重复键，后写入的键值对会覆盖前面的键值对；字典的值不要求唯一。


```python
my_dict['key2']  = 'value1'
print(my_dict)
```

字典可存储任意类型对象。


```python
my_dict = {'key1':123, 'key2':[12, 23, 33], 'key3':['item0','item1','item2']}
```

```python
print(my_dict['key3'])
```

```python
print(my_dict['key3'][0])
```

```python
print(my_dict['key3'][0].upper())
```

### 字典修改
字典是可变的，所以可以修改字典的值，如：


```python
my_dict['key1'] = 100
print(my_dict)
```

```python
my_dict['key1'] += 100
print(my_dict)
```

> 字典的值可以是任意数据类型；字典的键必须是可哈希对象，常见选择包括字符串、数字和元组。


我们还可以通过赋值创建键。例如，如果我们一开始有一个空字典，我们可以不断地添加:


```python
d = {}
# 赋值新增
d['animal'] = 'Dog'
d['grade'] = 'A'
print(d)
```

### 字典删除


字典既可以能删单一的元素，也能清空字典，清空只需一项操作。


```python
dict = {'Name': 'Zara', 'Age': 7, 'Class': 'First'}
print(dict.items())
del dict['Name']  # 删除键
dict.clear()  # 清空字典
print(dict)
del dict  # 删除变量
print(dict['Name']) # 会报错
print(dict['Age'])
```

### 字典遍历


- 遍历所有的键-值对


```python
favorite_language = {
    'jen' : 'python',
    'sarah' : 'c',
    'edward' : 'ruby',
    'phil' : 'python'
}

print(favorite_language)
for name, language in favorite_language.items():
    print(f"{name}'s favorite language is {language}")
```

注意：`items()` 方法可以按照迭代器的方式返回字典中的每一项。


- 遍历所有的键


如果不需要字典的值，而仅仅只是希望遍历字典的键时，可以使用 `keys()` 方法。


```python
for name in favorite_language.keys():
    print(name)
```

当然，如果不加 `keys()` 方法，其实遍历字典时，默认就是遍历字典的键。但是为了代码的可读性，还是建议显式地调用 `keys()` 方法更好。


```python
for name in favorite_language:
    print(name)
```

字典记录的是键和值之间的关联关系，字典中的元素顺序是不可预测的。当你想要以特定的顺序返回元素，一种办法是在 for 循环中对返回的键进行排序。为此，可使用函数 `sorted()` 来获得按特定顺序排列的键列表的副本。


```python
favorite_languages = {
'jen': 'python',
'sarah': 'c',
'edward': 'ruby',
'phil': 'python',
}
for name in sorted(favorite_languages.keys()):
    print(name.title())
for name, language in sorted(favorite_language.items(), key = lambda item:item[1]):
    print(name, language)
```

- 遍历字典中的所有值


如果你感兴趣的主要是字典包含的值，可使用方法 `values()` ，它返回一个值列表，而不包含任何键。例如上述例子中，如果你只关心涉及到的语言，那么可以：


```python
favorite_languages = {
'jen': 'python',
'sarah': 'c',
'edward': 'ruby',
'phil': 'python',
}
for language in favorite_languages.values():
    print(language.title())
```

上面的代码中，可以看到能够打印字典中所有的值。但是字典中只有键是唯一的，值并不唯一，也就可能存在重复的值，例如上面的 python 是重复。如果你不希望出现重复的值，可以将其转换成集合 set，那么每一个值都将是唯一的。


```python
for language in set(favorite_languages.values()):
    print(language.title())
```

### 字典常用方法


- `clear()` ：从字典中删除所有元素
- `copy()` ：返回字典的副本
- `get()` : 返回指定键的值
- `items()` : 返回包含每个键值对的元组的列表
- `keys()` : 返回一个包含字典键的列表
- `pop()` : 删除指定键的元素
- `popitem()` : 删除最后插入的键值对
- `update()` : 使用指定的键值对更新字典
- `values()` : 返回字典的所有值的列表


```python
# 字典方法
dict1 = {1: "Python", 2: "Java", 3: "Ruby", 4: "Scala"}

# copy()
dict2 = dict1.copy()
print(dict2)

# clear()
dict1.clear()
print(dict1)

# get()
print(dict2.get(1))

# items()
print(dict2.items())

# keys()
print(dict2.keys())

# pop()
dict2.pop(4)
print(dict2)

# popitem()
dict2.popitem()
print(dict2)

# update()
dict2.update({3: "Scala"})
print(dict2)

# values()
print(dict2.values())
```

## 集合


在 `Python` 中，集合 `set` 是无序、可迭代、可变且不包含重复元素的数据类型。

集合常用于去重、成员判断，以及数学集合运算，例如并集、交集和对称差集。与列表相比，集合的成员判断通常更高效。


### 集合创建


创建集合时，可以把元素放在花括号 `{}` 中，并用逗号分隔；也可以使用内置函数 `set()`。

集合可以包含不同类型的不可变元素，例如整数、浮点数、元组和字符串。但集合不能包含列表、集合或字典等可变对象。


```python
# 整数集合
my_set = {1, 2, 3}
print(my_set)

# 混合集合
my_set = {1.0, "Hello", (1, 2, 3)}
print(my_set)

# 字符串集合
my_set = set("python")
print(my_set)

# 列表创建集合
my_set = set([1, 2, 3])
print(my_set)

# 自动去重
set1 = set([1, 2, 4, 4, 3, 3, 3, 6, 5])
print(set1)
```

用可变元素创建集合会导致错误。


```python
set([1, 2, [3, 4]])
```

> 集合本身是可变对象，因此不能作为另一个集合的元素。


空花括号 `{}` 会创建空字典，而不是空集合。要创建空集合，应使用不带参数的 `set()`。


```python
empty_dict = {}
print(type(empty_dict))

empty_set = set()
print(type(empty_set))
```

### 集合修改


可以使用 `add()` 方法添加单个元素，可以通过循环添加多个。


```python
my_set = set()
print("Initial blank Set: ")
print(set1)

my_set.add(8)
my_set.add(9)
my_set.add((6, 7))
print("\nSet after Addition of Three elements: ")
print(my_set)

for i in range(1, 6):
    my_set.add(i)
print("\nSet after Addition of elements from 1-5: ")
print(my_set)
```

也可以使用 `update()` 方法添加多个元素。`update()` 方法可以接受元组、列表、字符串或其他集合作为参数。在任何情况下，都要避免重复。


```python
my_set = set([4, 5, (6, 7)])
my_set.update([10, 11])
print("\nSet after Addition of elements using Update: ")
print(my_set)
```

虽然集合是可变的，但集合不能通过引用索引来访问，因为集合是无序的，所以集合没有索引。但是可以使用 for 循环遍历集合，或者通过使用 in 关键字查询集合中是否存在指定的值。


```python
my_set = set(["python", "java", "ruby"])
print("\nInitial set")
print(my_set)

# 遍历集合
print("\nElements of set: ")
for i in my_set:
    print(i, end=" ")

print("\n")
# 成员判断
print("python" in my_set)
```

### 集合删除


可以使用 `discard()` 和 `remove()` 从集合中删除指定元素。

两者区别在于：如果元素不存在，`discard()` 会保持集合不变；`remove()` 会抛出 `KeyError`。


```python
# 初始化集合
my_set = {1, 3, 4, 5, 6}
print(my_set)

# discard 删除
my_set.discard(4)
print(my_set)

# remove 删除
my_set.remove(6)
print(my_set)

# discard 忽略缺失值
my_set.discard(2)
print(my_set)

# remove 缺失值会报错
# KeyError
my_set.remove(2)
```

类似地，我们可以使用 `pop()` 方法删除并返回一个项。由于 set 是无序数据类型，因此无法确定将弹出哪个项。这完全是随意的。

还可以使用 `clear()` 方法从集合中删除所有项。


```python
my_set = set("HelloWorld")
print(my_set)

# 弹出元素
print(my_set.pop())

# 再次弹出
my_set.pop()
print(my_set)

# 清空集合
my_set.clear()
print(my_set)
```

### 集合运算


集合可以用来执行数学集合操作，如并集、交集、差集和对称差集。


```python
# 并集
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

print(A | B)
print(A.union(B))
print(B.union(A))
```

```python
# 交集
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

print(A & B)
print(A.intersection(B))
print(B.intersection(A))
```

```python
# 补集
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

print(A - B)
print(A.difference(B))
print(B.difference(A))
```

```python
# 对称差
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

print(A ^ B)
print(A.symmetric_difference(B))
print(B.symmetric_difference(A))
```

### 冻结集合 `frozenset`


Frozenset 是一个新的类，它具有集合的特征，但它的元素一旦被分配就不能被更改。元组是不可变列表，frozenset 是不可变集。

可变的集合是不可哈希的，因此它们不能用作字典键。另一方面，冻结集是可哈希的，可以用作字典的键。

该数据类型支持 `copy()`, `difference()`, `intersection()`, `isdisjoint()`, `issubset()`, `issuperset()`, `symmetric_difference()` 和 `union()` 等方法。由于是不可变的，它没有添加或删除元素的方法。


`frozenset可以使用frozenset()` 函数创建。


```python
A = frozenset([1, 2, 3, 4])
B = frozenset([3, 4, 5, 6])
A.isdisjoint(B)
```

```python
frozenset({1, 2, 3, 4, 5, 6})
A.add(3) # 不可变，会报错
```

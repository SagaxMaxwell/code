## 列表基础


前面讨论字符串时，我们介绍了 `Python` 中序列 `sequence` 的概念。列表是最常用的序列类型之一。与字符串不同，列表是可变对象，列表中的元素可以被修改。

本节主要讨论：

1. 创建列表。
2. 索引和切片。
3. 常用列表方法。
4. 遍历列表。
5. 嵌套列表。
6. 列表推导式。


列表由方括号\[\]和分隔列表中的每个元素的逗号组成。


```python
my_list = [1, 2, 3]
```

上面我们创建了一个全部由整数组成的列表。事实上，列表中的元素可以不同类型，列如：


```python
my_list = ["A string",1, 3.14, 1+1j]
print(my_list)
```

和字符串一样，`len()` 函数可以获取列表序列中有多少项。


```python
print(len(my_list))
```

## 索引与切片


列表的索引与切片基本与字符串的操作一致。


```python
my_list = ['one', 'two','three',4, 5]
```

```python
#抽取第一个元素
print(my_list[0])
```

```python
#抽取索引1开始的所有元素
print(my_list[1:])
```

```python
#抽取从头开始到第3个元素
print(my_list[:3])
```

了解列表的切片操作之后，那么复制一个列表就非常简单了。


与字符串一样，可以通过运算符‘+’来拼接不同的列表，例如


```python
print(my_list + ['add new item'])
```

当然，上面的操作并不会改变原来列表的内容，打印 my_list 的内容，应该会看到和原来是一样的。


```python
print(my_list)
```

可以通过重新赋值的方式改变列表的值，例如：


```python
my_list = my_list + ["append a new item"] # 等价于 my_list += ["append a new item"]
print(my_list)
```

类似于字符串，可以使用\*符号复制列表，例如：


```python
print(my_list * 2)
```

## 列表方法


`Python` 的列表与 `C` / `C++` 数组有相似之处，但操作更灵活。列表没有固定大小，也不要求所有元素类型一致，因此更接近动态数组或泛型容器。


```python
my_list = [1, 2, 3]
```

- **append**方法在列表末尾添加元素。


```python
my_list.append("append a new item")
print(my_list)
```

- **extend**方法通过迭代的方式在列表末尾添加元素。注意与 append 方法的区别。例如


```python
my_list = [1, 2, 3]
my_list.append([4, 5])
print(my_list)
my_list = [1, 2, 3]
my_list.extend([4, 5])
print(my_list)
```

- **count**接受一个元素并返回它在列表中出现的次数。


```python
my_list = [1, 2, 3, 4, 3, 4, 5]
print(my_list.count(3))
```

- **index**将返回作为参数放置的任何元素的索引。注意:如果元素不在列表中，则返回错误。


```python
my_list = [0, 1, 2, 3, 4]
my_list.index(2)
```

- **pop**方法删除列表元素。


使用 pop 从列表中删除一个元素。默认情况下 pop 会删除最后一个索引，但是你也可以指定删除哪个索引。例如:


```python
my_list = [1, 2, 3, 4]
my_list.pop() #默认删除最后一个元素
print(my_list)
```

```python
my_list.pop(0) #删除指定索引的元素
print(my_list)
```

- **remove**删除第一个符合要求的元素


```python
my_list = [1, 2, 3, 4, 3]
my_list.remove(3)
print(my_list)
```

- **`sort`**和**reverse**方法分别将列表中元素排序和逆序。


```python
my_list = ['y', 'e', 'm', 'e', 'n', 'g', 'j', 'i', 'e']
print(my_list)
```

```python
my_list.sort()
print(my_list)
```

## 遍历列表


在很多的应用场景，我们需要遍历整个列表，对每一个元素执行相同的操作。遍历列表通常常用 for 循环来实现。例如：


```python
courses = ['C++', 'Python', "Java"]
for course in courses:
    print(course)
```

遍历列表需要注意的问题是:
- for 循环后面不要忘记加上冒号:
- for 循环里面的语句缩进要小心，相同的缩进表示都在循环里面。


掌握了遍历列表的方法后，我们创建列表经常使用的方法是采用 range 方法。`range()` 可以很方便地生成一系列的数字。例如：


```python
for value in range(1, 11):
    print(value, end=' ')
```

使用 `range()` 方法时，注意 `range(a,b)` 表示的区间时\[a,b)，即左闭右开区间。


另外，还要注意，`range()` 可以产生一系列数字，但它返回的类型并不是列表，可以打印一下试一试。


```python
range_list = range(1, 11)
print(type(range_list))
```

但时我们可以使用 list 方法将 range 类型转换为列表。


```python
numbers = list(range(1, 11))
print(numbers)
print(type(numbers))
```

## 嵌套列表


```python
my_list.reverse()
print(my_list)
```

`Python` 数据结构支持嵌套。这意味着可以在一种数据结构中存放另一种数据结构，例如列表中的列表。


```python
# Let's make three lists
lst_1=[1, 2, 3]
lst_2=[4, 5, 6]
lst_3=[7, 8, 9]

# Make a list of lists to form a matrix
matrix = [lst_1, lst_2, lst_3]
print(matrix)
```

现在我们可以再次使用索引来获取元素，但是现在索引有两个级别。矩阵对象中的项，然后是列表中的项!


```python
print(matrix[0])
print(matrix[0][0])
```

## 列表推导


列表推导式是 `Python` 中快速构建列表的常用语法。它可以把一个简单循环和可选条件写在方括号内，让列表生成逻辑更紧凑。

下面通过几个例子理解列表推导式。


### 示例 1


```python
#创建一个字符串
my_string = "something"
print(my_string)
#用字符串的每一个字符构成列表
my_list = [c for c in my_string]
print(my_list)
```

### 示例 2


```python
#区间[0, 11)的平方数
my_list = [x**2 for x in range(0, 11)]
print(my_list)
```

### 示例 3


```python
#区间[0, 11)的所有偶数
my_list = [x for x in range(11) if x % 2 == 0]
print(my_list)
```

### 示例 4


```python
# Convert Celsius to Fahrenheit
celsius = [0, 10, 20.1, 34.5]

fahrenheit = [ ((float(9)/5)*temp + 32) for temp in celsius ]

print(fahrenheit)
```

### 示例 5


使用嵌套推导


```python
my_list = [x**2 for x in [x**2 for x in range(0, 11)]]
print(my_list)
```

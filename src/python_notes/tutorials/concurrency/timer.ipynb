# threading.Timer

`Timer` 用于延迟启动一个函数调用。本质上它仍然会创建线程，适合轻量延迟任务，不适合复杂任务调度。


```python
import threading
import time
from typing import Any


def do_something(name: Any, queue: list[Any]) -> None:
    queue.append(name)


pool: list[Any] = []
timer = threading.Timer(2, do_something, args=('Alice', pool))
timer.start()

while not pool:
    print('Waiting for Alice...')
    time.sleep(1)

print(pool)
```

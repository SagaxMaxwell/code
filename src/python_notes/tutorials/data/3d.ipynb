# Matplotlib 三维图形

Matplotlib 通过 `mplot3d` 支持三维坐标轴。三维图适合展示空间曲线、散点、三维柱形、轮廓、框线和表面。

## 三维坐标轴

使用 `projection="3d"` 创建三维 Axes。


```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d


fig = plt.figure()
ax = plt.axes(projection="3d")
```

## 三维点和线

`plot3D()` 绘制三维折线，`scatter3D()` 绘制三维散点。


```python
rng = np.random.default_rng(42)
ax = plt.axes(projection="3d")
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, "gray")

zdata = 15 * rng.random(100)
xdata = np.sin(zdata) + 0.1 * rng.standard_normal(100)
ydata = np.cos(zdata) + 0.1 * rng.standard_normal(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap="Dark2")
```

## 三维柱形图

`bar3d()` 使用三维位置和长宽高绘制柱体。


```python
ax = plt.axes(projection="3d")
x = np.arange(10)
y = np.arange(10)
z = np.zeros(10)
dx = np.ones(10)
dy = np.ones(10)
dz = np.arange(10)
ax.bar3d(x, y, z, dx, dy, dz, color="b", alpha=0.8)
```

## 三维轮廓和表面

网格数据可以绘制三维轮廓图、框线图和表面图。


```python
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))


x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection="3d")
ax.contour3D(X, Y, Z, 50, cmap="binary")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.view_init(60, 35)
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot_wireframe(X, Y, Z, color="black")
ax.set_title("wireframe")

fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap="viridis", edgecolor="none")
ax.set_title("surface")
```

# `Selenium` 入门

**学习目标：** 本节介绍 `Selenium 4` 的环境准备、元素定位、浏览器控制、鼠标键盘事件、窗口框架切换和下拉菜单操作。


## 什么是 `Selenium`


`Selenium` 是一套用于自动化浏览器操作的工具，常用于 `Web` 自动化测试、端到端测试、页面数据采集和重复性浏览器任务。

在 `Python` 中，常用的是 `Selenium WebDriver`。它可以启动真实浏览器，访问页面，定位元素，并执行输入、点击、切换窗口、切换框架等操作。

> **注意：** 本教程按 `Selenium 4` 的现代 API 编写。元素定位统一使用 `find_element(By.ID, "...")`、`find_element(by=By.ID, value="...")` 或对应的 `find_elements()` 写法，不使用旧版定位快捷方法。

大多数本地学习场景可以先使用 `webdriver.Chrome()` 或 `webdriver.Firefox()` 让 `Selenium Manager` 自动处理驱动。


> **注意：** 如果手动管理浏览器驱动，需要将驱动放入系统 `PATH`，或者通过对应浏览器的 `Service` 对象显式指定驱动路径。


基础启动示例：


````python
from selenium import webdriver
from selenium.webdriver.common.by import By

# 启动浏览器并打开官方测试表单页面。
driver = webdriver.Chrome()
driver.get("https://www.selenium.dev/selenium/web/web-form.html")
````

## 元素定位


`Selenium 4` 统一使用 `find_element()` 和 `find_elements()` 定位元素，并通过 `By` 指定定位策略。

| 定位一个元素 | 定位多个元素 | 含义 |
| :--- | :--- | :--- |
| `find_element(By.ID, "my-text-id")` | `find_elements(By.ID, "my-text-id")` | 通过 `id` 定位。 |
| `find_element(By.NAME, "my-text")` | `find_elements(By.NAME, "my-text")` | 通过 `name` 定位。 |
| `find_element(By.CLASS_NAME, "form-control")` | `find_elements(By.CLASS_NAME, "form-control")` | 通过类名定位。 |
| `find_element(By.TAG_NAME, "input")` | `find_elements(By.TAG_NAME, "input")` | 通过标签名定位。 |
| `find_element(By.LINK_TEXT, "Return to index")` | `find_elements(By.LINK_TEXT, "Return to index")` | 通过完整链接文本定位。 |
| `find_element(By.PARTIAL_LINK_TEXT, "index")` | `find_elements(By.PARTIAL_LINK_TEXT, "index")` | 通过部分链接文本定位。 |
| `find_element(By.XPATH, "//input[@name='my-text']")` | `find_elements(By.XPATH, "//input")` | 通过 `XPath` 定位。 |
| `find_element(By.CSS_SELECTOR, "#my-text-id")` | `find_elements(By.CSS_SELECTOR, "input")` | 通过 `CSS Selector` 定位。 |


````python
# 通过 `id` 定位。
text_by_id = driver.find_element(by=By.ID, value="my-text-id")
print(text_by_id)
````

````python
# 通过 `name` 定位。
text_by_name = driver.find_element(by=By.NAME, value="my-text")
print(text_by_name)
````

````python
# 通过 `class name` 定位。
text_by_class_name = driver.find_element(by=By.CLASS_NAME, value="form-control")
print(text_by_class_name)
````

````python
# 通过标签名定位。
first_input = driver.find_element(by=By.TAG_NAME, value="input")
print(first_input)
````

通过 `XPath` 定位时，可以根据属性、层级和条件表达式编写定位规则。


````python
print(len(driver.find_elements(by=By.XPATH, value="//input")))
print(driver.find_element(by=By.XPATH, value="//*[@id='my-text-id']"))
print(driver.find_element(by=By.XPATH, value="//*[@name='my-text']"))
print(driver.find_element(by=By.XPATH, value="//input[@class='form-control']"))
print(driver.find_element(by=By.XPATH, value="//input[@id='my-text-id' and @name='my-text']"))
````

通过 `CSS Selector` 定位时，可以使用 `id`、属性、类名和标签组合等写法。


````python
print(driver.find_element(by=By.CSS_SELECTOR, value="#my-text-id"))
print(driver.find_element(by=By.CSS_SELECTOR, value="[name='my-text']"))
print(driver.find_element(by=By.CSS_SELECTOR, value=".form-control"))
print(driver.find_element(by=By.CSS_SELECTOR, value="input#my-text-id"))
````

根据链接文本定位元素，适合定位页面中的超链接。


````python
driver.get("https://www.selenium.dev/documentation/")
print(driver.find_element(by=By.LINK_TEXT, value="WebDriver"))
print(driver.find_element(by=By.PARTIAL_LINK_TEXT, value="Web"))
````

## 浏览器控制


| 方法 | 说明 |
| :--- | :--- |
| `driver.set_window_size(480, 800)` | 设置浏览器窗口大小。 |
| `driver.maximize_window()` | 最大化浏览器窗口。 |
| `driver.back()` | 控制浏览器后退。 |
| `driver.forward()` | 控制浏览器前进。 |
| `driver.refresh()` | 刷新页面。 |
| `driver.find_element(By.ID, "my-text-id").clear()` | 清除文本。 |
| `driver.find_element(By.ID, "my-text-id").send_keys("Selenium")` | 模拟键盘输入。 |
| `driver.find_element(By.CSS_SELECTOR, "button").click()` | 单击元素。 |
| `driver.find_element(By.CSS_SELECTOR, "button").submit()` | 提交表单。 |
| `element.get_attribute("name")` | 获取元素属性值。 |
| `element.is_displayed()` | 判断元素是否可见。 |
| `element.size` | 返回元素大小。 |
| `element.text` | 获取元素文本。 |


下面是打开官方测试表单页面，并完成输入和提交的示例。


````python
driver.get("https://www.selenium.dev/selenium/web/web-form.html")
driver.refresh()

driver.set_window_size(1400, 800)
driver.maximize_window()

print(driver.current_window_handle)

text_box = driver.find_element(by=By.ID, value="my-text-id")
text_box.clear()
text_box.send_keys("Selenium")

submit_button = driver.find_element(by=By.CSS_SELECTOR, value="button")
submit_button.click()
````

````python
print(driver.window_handles)
message = driver.find_element(by=By.ID, value="message")
print(message.text)
````

### 定位一组元素


````python
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep

# 打开官方文档页面，并定位页面中的链接元素。
driver = webdriver.Chrome()
driver.get("https://www.selenium.dev/documentation/")
driver.implicitly_wait(0.5)

links = driver.find_elements(by=By.CSS_SELECTOR, value="a")
print(type(links))

for link in links[:10]:
    text = link.text.strip()
    if text:
        print(text)

sleep(0.1)
````

## 鼠标操作


在 `WebDriver` 中，鼠标操作由 `ActionChains` 类封装。

`ActionChains` 常用方法包括：

1. `perform()`：执行当前链式操作中存储的所有行为。
2. `context_click()`：右击。
3. `double_click()`：双击。
4. `drag_and_drop()`：拖动。
5. `move_to_element()`：鼠标悬停。


````python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By

# 打开官方文档页面。
driver = webdriver.Chrome()
driver.get("https://www.selenium.dev/documentation/")
driver.maximize_window()

# 定位到导航链接，并执行鼠标悬停。
element = driver.find_element(by=By.LINK_TEXT, value="WebDriver")
ActionChains(driver).move_to_element(element).perform()
````

## 键盘事件


以下为常用键盘操作。需要先导入 `Keys`：

```python
from selenium.webdriver.common.keys import Keys
```

- `element.send_keys(Keys.BACK_SPACE)`：删除键 `Backspace`。
- `element.send_keys(Keys.SPACE)`：空格键 `Space`。
- `element.send_keys(Keys.TAB)`：制表键 `Tab`。
- `element.send_keys(Keys.ESCAPE)`：退出键 `Esc`。
- `element.send_keys(Keys.ENTER)`：回车键 `Enter`。
- `element.send_keys(Keys.CONTROL, "a")`：全选，等价于 `Ctrl+A`。
- `element.send_keys(Keys.CONTROL, "c")`：复制，等价于 `Ctrl+C`。
- `element.send_keys(Keys.CONTROL, "x")`：剪切，等价于 `Ctrl+X`。
- `element.send_keys(Keys.CONTROL, "v")`：粘贴，等价于 `Ctrl+V`。
- `element.send_keys(Keys.F1)` 到 `element.send_keys(Keys.F12)`：功能键 `F1` 到 `F12`。


## 窗口与框架切换


窗口和框架切换分别使用 `driver.switch_to.window()` 与 `driver.switch_to.frame()`。前者接收窗口句柄，后者可以接收框架名称、索引或已经定位到的 `iframe` 元素。


````python
driver.switch_to.window("window_name")
driver.switch_to.frame("frame_name")
````

可以直接使用 `iframe` 的 `id` 或 `name` 属性切换框架。如果 `iframe` 没有可用的 `id` 或 `name`，可以先定位到 `iframe` 元素，再传给 `driver.switch_to.frame()`。


如果 `iframe` 没有稳定的 `id` 或 `name`，先用 `driver.find_element()` 定位到 `iframe` 元素，再调用 `driver.switch_to.frame()` 切换上下文。


````python
iframe = driver.find_element(by=By.XPATH, value="//*[@id='x-URS-iframe']")
driver.switch_to.frame(iframe)
````

完成 `frame` 中的操作后，可以返回默认页面上下文。


完成 `frame` 内部操作后，调用 `driver.switch_to.default_content()` 返回默认页面上下文。


````python
driver.switch_to.default_content()
````

## 下拉菜单


导入下拉框选择类 `Select` 后，可以用它处理 `<select>` 下拉框。

常用方法包括：

- `select_by_value("2")`：按 `value` 属性选择。
- `select_by_visible_text("Two")`：按可见文本选择。
- `select_by_index(1)`：按选项索引选择。


````python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.select import Select
from time import sleep

# 打开官方测试表单页面。
driver = webdriver.Chrome()
driver.implicitly_wait(10)
driver.get("https://www.selenium.dev/selenium/web/web-form.html")
sleep(0.5)
````

````python
select_element = driver.find_element(by=By.NAME, value="my-select")
select = Select(select_element)
select.select_by_value("2")
````

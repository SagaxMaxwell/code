# 双指针

## 合并两个有序数组

### 注意事项

- 空列表
- 数组越界
- 不同数据对象指针

```python
class Solution:
    def merge(self, nums1: list[int], m: int, nums2: list[int], n: int) -> None:
        """
        Do not return anything, modify nums1 in-place instead.
        """
        result = list()
        i, j = 0, 0

        if n == 0:
            nums1[:] = nums1[:m]
            return
        elif m == 0:
            nums1[:] = nums2[:n]
            return

        while True:
            if nums1[i] <= nums2[j]:
                result.append(nums1[i])
                i += 1
                if i == m:
                    result.extend(nums2[j:n])
                    break
            else:
                result.append(nums2[j])
                j += 1
                if j == n:
                    result.extend(nums1[i:m])
                    break

        nums1[:] = result[:]


nums1 = []
m = 0
nums2 = [1, 4]
n = 2

s = Solution()
s.merge(nums1, m, nums2, n)

nums1
```

## 移除元素

### 注意事项

- 空列表
- 数组越界
- 相邻指针

```python
class Solution:
    def removeElement(self, nums: list[int], val: int) -> int:
        left = 0

        for right in range(len(nums)):
            if nums[right] != val:
                nums[left] = nums[right]
                left += 1

        return left


nums = [3, 2, 2, 3]
val = 3

s = Solution()
s.removeElement(nums, val)
```

## 删除有序数组中的重复项

### 注意事项

- 空列表
- 数组越界
- 快慢指针

```python
class Solution:
    def removeDuplicates(self, nums: list[int]) -> int:
        slow = 0

        for fast in range(len(nums)):
            if nums[fast] != nums[slow]:
                slow += 1
                nums[slow], nums[fast] = nums[fast], nums[slow]

        return slow + 1


nums = [0, 0, 1, 1, 1, 2, 2, 3, 3, 4]

s = Solution()
s.removeDuplicates(nums)
```

## 删除有序数组中的重复项 II

### 注意事项

- 空列表
- 数组越界
- 快慢指针 + 步长变动
- slow 不是普通指针，而是答案区的长度
- nums[0:slow] 始终是当前已经构造好的合法结果
- 我不是真的删除元素，我只需要把应该保留的元素依次覆盖到数组前面
- 建模
    - 答案：nums 的前 k 个元素
    - 合法标准：每个数字最多出现 2 次
    - 状态：nums[:slow] 是已经构造好的合法答案前缀
    - 不变量：nums[:slow] 始终合法
    - 当前元素：nums[fast]
    - 判断：nums[fast] != nums[slow - 2]
    - 转移：能加入就写到 nums[slow]，slow += 1
    - 顺序：fast 从左到右扫描
    - 返回：slow

```python
class Solution:
    def removeDuplicates(self, nums: list[int]) -> int:
        slow, fast = 2, 2
        n = len(nums)

        if n < 2:
            return n

        while fast < n:
            if nums[fast] != nums[slow - 2]:
                nums[fast], nums[slow] = nums[slow], nums[fast]
                slow += 1
            fast += 1

        return slow


nums = [0, 0, 1, 1, 1, 1, 2, 3, 3]

s = Solution()
s.removeDuplicates(nums)
```

## 多数元素

### 注意事项

- 哈希表
- 答案是什么（先确定最终要返回/构造的东西）
    - 变量 num
- 合法标准是什么（什么样的答案算对）
    - num 出现的次数 > n // 2
- 我维护什么状态（算法中间过程手里保存什么）
    - 元素出现次数的字典
- 状态的不变量是什么（每一步之后，什么必须一直成立）
    - 扫描完 nums[:i] 后，字典记录了 nums[:i] 中每个元素的真实出现次数
- 当前元素/选择怎么处理（能不能加入？加入条件是什么？不加入怎么办？）
    - 当前元素为 nums[i]
- 状态怎么转移（指针怎么动、数组怎么写、dp 怎么更新、队列/栈怎么变化）
    - 在哈希表里将 nums[i] 的出现次数 ++
    - i++   
- 处理顺序是什么（从左到右？从右到左？先排序？按层？递归？二分？）
    - i 从左到右扫表
    - step 1
- 初始化和返回值是什么（起点、边界、最终返回哪个变量）
    - num

```python
from collections import Counter


class Solution:
    def majorityElement(self, nums: list[int]) -> int:
        num, count = 0, 0

        for k, v in Counter(nums).items():
            if v >= count:
                num, count = k, v

        return num


nums = [3, 2, 3]

s = Solution()
s.majorityElement(nums)
```

## 多数元素

### 注意事项

- 候选人
- 答案是什么（先确定最终要返回/构造的东西）
    - 变量 num
- 合法标准是什么（什么样的答案算对）
    - num 的出现次数 > n // 2
- 我维护什么状态（算法中间过程手里保存什么）
    - 元素与出现次数的字典
    - num（当前出现最多次的元素）
    - count（num 当前出现的次数）
- 状态的不变量是什么（每一步之后，什么必须一直成立）
    - 扫描完 nums[:i] 后，字典记录了 nums[:i] 中每个元素的真实出现次数
    - count 是 nums[:i] 中目前最大的出现次数
    - num 是某个出现次数等于 count 的元素
- 当前元素/选择怎么处理（能不能加入？加入条件是什么？不加入怎么办？）
    - 当前元素为 nums[i]
- 状态怎么转移（指针怎么动、数组怎么写、dp 怎么更新、队列/栈怎么变化）
    - 字典将 nums[i] 的出现次数 ++
    - i++
    - 字典里找当前元素出现的次数 >= count 就更新 num = nums[i]
- 处理顺序是什么（从左到右？从右到左？先排序？按层？递归？二分？）
    - i 从左到右扫表
    - step 1
- 初始化和返回值是什么（起点、边界、最终返回哪个变量）
    - num

```python
from collections import defaultdict


class Solution:
    def majorityElement(self, nums: list[int]) -> int:
        counts = defaultdict(lambda: 0)
        num, count = 0, 0

        for v in nums:
            counts[v] += 1
            if counts[v] >= count:
                num, count = v, counts[v]

        return num


nums = [3, 2, 3]

s = Solution()
s.majorityElement(nums)
```

## 轮转数组

### 注意事项

- 数组越界
- 空列表
- 答案是什么（先确定最终要返回/构造的东西）
    - 原地排序完的数组
- 合法标准是什么（什么样的答案算对）
    - 元素不变，但数组元素顺序向右轮转 k 位
- 我维护什么状态（算法中间过程手里保存什么）
    - _ 向右轮转的次数
    - nums[idx] 迭代的元素
    - idx 当前的索引
    - nums 排序完的数组
- 状态的不变量是什么（每一步之后，什么必须一直成立）
    - 数组元素不变
- 当前元素/选择怎么处理（能不能加入？加入条件是什么？不加入怎么办？）
    - 当前元素为 nums[idx]
- 状态怎么转移（指针怎么动、数组怎么写、dp 怎么更新、队列/栈怎么变化）
    - nums[idx] 与 nums[idx-1] 调换
- 处理顺序是什么（从左到右？从右到左？先排序？按层？递归？二分？）
    - 从右到左
    - step -1
- 初始化和返回值是什么（起点、边界、最终返回哪个变量）
    - 返回值是无

```python
class Solution:
    def rotate(self, nums: list[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        for _ in range(k):
            for idx in range(len(nums) - 1, 0, -1):
                nums[idx], nums[idx - 1] = nums[idx - 1], nums[idx]

nums = [1, 2]
k = 3


s = Solution()
s.rotate(nums, k)
```

## 轮转数组

### 注意事项

- 答案是什么（先确定最终要返回/构造的东西）
    - 原地排序完的数组
- 合法标准是什么（什么样的答案算对）
    - 元素不变，但数组元素顺序向右轮转 k 位
- 我维护什么状态（算法中间过程手里保存什么）
    - 处理过冗余次数的 k
    - nums 排序完的数组
- 状态的不变量是什么（每一步之后，什么必须一直成立）
    - 数组元素不变
- 当前元素/选择怎么处理（能不能加入？加入条件是什么？不加入怎么办？）
    - nums[-k:]
    - nums[:-k]
- 状态怎么转移（指针怎么动、数组怎么写、dp 怎么更新、队列/栈怎么变化）
    - nums = nums[:-k] = nums[-k:]
- 处理顺序是什么（从左到右？从右到左？先排序？按层？递归？二分？）
    - 从右到左
- 初始化和返回值是什么（起点、边界、最终返回哪个变量）
    - 返回值是无

```python
class Solution:
    def rotate(self, nums: list[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        k %= len(nums)
        nums[:] = nums[-k:] + nums[:-k]


nums = [1, 2]
k = 3

s = Solution()
s.rotate(nums, k)
```

## 买卖股票最佳时机

### 注意事项

- 数组越界
- 空列表
- 答案是什么（先确定最终要返回/构造的东西）
    - prices[stop] - prices[start]
- 合法标准是什么（什么样的答案算对）
    - stop >= start
    - prices[stop] - prices[start] 的最大值
- 我维护什么状态（算法中间过程手里保存什么）
    - start
    - stop
    - current
- 状态的不变量是什么（每一步之后，什么必须一直成立）
    - 数组不变
    - prices[stop] - prices[start] >= 0
- 当前元素/选择怎么处理（能不能加入？加入条件是什么？不加入怎么办？）
    - prices[start]
    - prices[stop]
    - prices[current]
- 状态怎么转移（指针怎么动、数组怎么写、dp 怎么更新、队列/栈怎么变化）
    - if current > start and prices[current] - prices[start] > prices[stop] - prices[start]: stop = current
    - if current > start and prices[current] < prices[start]: start = current
- 处理顺序是什么（从左到右？从右到左？先排序？按层？递归？二分？）
    - 从左到右
    - step 1
- 初始化和返回值是什么（起点、边界、最终返回哪个变量）
    - prices[stop] - prices[start]

```python
class Solution:
    def maxProfit(self, prices: list[int]) -> int:
        buy = 0
        sell = 0
        profit = 0
        min = 0

        for fast in range(len(prices)):
            if prices[fast] - prices[buy] > profit:
                profit = prices[fast] - prices[buy]
                sell = fast

            if prices[fast] - prices[min] > profit:
                buy, sell = min, fast
                profit = prices[fast] - prices[min]

            if prices[fast] < prices[min]:
                min = fast

        return prices[sell] - prices[buy]


prices = [7, 1, 4, 2]

s = Solution()
s.maxProfit(prices)
```

# 双指针

双指针常用于有序数组、区间收缩和原地处理问题。核心是每轮移动哪个指针，以及移动后是否仍保持不变量。

## 两数之和

在升序数组中找到两个数，使其和等于目标值。


```python
def two_sum(nums: list[int], target: int) -> list[int]:
    left = 0
    right = len(nums) - 1

    while left < right:
        current = nums[left] + nums[right]
        if current == target:
            return [left + 1, right + 1]
        if current > target:
            right -= 1
        else:
            left += 1

    return []


two_sum([-1000, -1, 0, 1], 1)
```

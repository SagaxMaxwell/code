## Sliding Window（滑动窗口）

### 无重复字符的最长子串

- 给定字符串 `s`，返回不含重复字符的最长子串长度

```python
def length_of_longest_substring(s):
    start = 0
    max_len = 0
    char_set = set()

    for end in range(len(s)):
        while s[end] in char_set:
            char_set.remove(s[start])
            start += 1
        char_set.add(s[end])
        max_len = max(max_len, end - start + 1)

    return max_len


input_string = "abcbbabcbb"
print(length_of_longest_substring(input_string))
```

## Modified Binary Search（变形二分查找）

### 在旋转排序数组中搜索

- `nums` 是按升序排列且元素互异的整数数组
- `nums` 可能在传入函数前按某个未知下标做过旋转

```python
def search(nums: list, target: int) -> int:
    left, right = 0, len(nums) - 1

    while left <= right:
        mid = (left + right) // 2

        if nums[mid] == target:
            return mid
        elif nums[left] <= nums[mid]:
            if nums[left] <= target < nums[mid]:
                right = mid - 1
            else:
                left = mid + 1
        else:
            if nums[mid] < target <= nums[right]:
                left = mid + 1
            else:
                right = mid - 1

    return -1


nums = [4,5,6,7,0,1,2]
target = 0

search(nums, target)
```

## Top K Elements（前 K 个元素）

### 数组中第 `K` 大元素

- 给定整数数组 `nums` 和整数 `k`，返回数组中的第 `k` 大元素

```python
import heapq


# 排序法
def find_kth_largest(nums: list, k: int) -> int:
    nums.sort(reverse=True)
    return nums[k-1]


# 堆选择法
def find_kth_largest_with_heap(nums: list, k: int) -> int:
    return heapq.nlargest(k, nums)[-1]


nums = [3,2,1,5,6,4]
k = 2

print(find_kth_largest(nums, k))
print(find_kth_largest_with_heap(nums, k))
```

## Binary Tree DFS（二叉树深度优先遍历）

### DFS（前序遍历）

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


# 递归实现
def preorder_traversal(node):
    if node is not None:
        print(node.val)
        preorder_traversal(node.left)
        preorder_traversal(node.right)


# 迭代实现
def preorder_traversal_loop(root):
    result = []
    stack = [root]

    while stack:
        node = stack.pop()
        result.append(node.val)

        if node.right:
            stack.append(node.right)
        if node.left:
            stack.append(node.left)

    return result


# 构建一个简单的二叉树
#       1
#      / \
#     2   3
#    / \
#   4   5
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)


preorder_traversal(root)
preorder_traversal_loop(root)
```

### DFS（中序遍历）

```python
# 递归实现
def inorder_traversal(node):
    if node is not None:
        inorder_traversal(node.left)
        print(node.val)
        inorder_traversal(node.right)


# 迭代实现
def inorder_traversal_loop(root):
    result = []
    stack = []
    current = root

    while current or stack:
        while current:
            stack.append(current)
            current = current.left

        current = stack.pop()
        result.append(current.val)
        current = current.right

    return result


inorder_traversal(root)
inorder_traversal_loop(root)
```

### DFS（后序遍历）

```python
# 递归实现
def postorder_traversal(node):
    if node is not None:
        postorder_traversal(node.left)
        postorder_traversal(node.right)
        print(node.val)


# 迭代实现
def postorder_traversal_loop(root):
    result = []
    stack = [root]

    while stack:
        node = stack.pop()
        result.append(node.val)

        if node.left:
            stack.append(node.left)
        if node.right:
            stack.append(node.right)

    return result[::-1]


# 反转前序遍历结果（仅作对比，不等同于后序遍历）
def reverse_preorder_traversal(root):
    return preorder_traversal_loop(root)[::-1]


postorder_traversal(root)
print(postorder_traversal_loop(root))
print(reverse_preorder_traversal(root))
```

### 计算二叉树最大深度

```python
def tree_depth(node):
    if node is None:
        return 0

    left_depth = tree_depth(node.left)
    right_depth = tree_depth(node.right)

    return max(left_depth, right_depth) + 1


tree_depth(root)
```

### BFS（层序遍历）

```python
def bfs_binary_tree(root):
    queue = [root]
    result = []

    while queue:
        node = queue.pop(0)
        result.append(node.val)

        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)

    return result


bfs_binary_tree(root)
```

## Graph Algorithms（图算法）

### 单源最短路径

- Dijkstra 算法

```python
def dijkstra(graph, node):
    S = {node: 0}
    U = dict({point: float("infinity") for point in graph}, **graph.get(node))
    U.pop(node, None)

    while len(U) > 0:
        U = dict(sorted(U.items(), key=lambda item: item[1], reverse=True))
        cur_point, cur_dis = U.popitem()
        S.update({cur_point: cur_dis})

        for point, dis in graph.get(cur_point).items():
            if U.get(point) and U.get(point) >= dis + cur_dis:
                U.update({point: dis + cur_dis})

    return S


graph = {
    'A': {'B': 10, 'C': 3},
    'B': {'A': 10, 'C': 1, 'D': 2},
    'C': {'A': 3, 'B': 1, 'D': 4},
    'D': {'B': 2, 'C': 4}
}

distances = dijkstra(graph, 'A')
print(distances)
```

### 图 BFS（广度优先搜索）

```python
def bfs_graph(graph, start):
    result = []
    visited = {start}
    queue = [start]

    while len(queue) > 0:
        point = queue.pop(0)
        result.append(point)

        for next_point in graph.get(point, []):
            if next_point not in visited:
                visited.add(next_point)
                queue.append(next_point)

    return result


graph = {
    'A': ['B', 'C', 'D'],
    'B': ['A', 'E'],
    'C': ['A', 'F'],
    'D': ['A', 'G'],
    'E': ['B', 'H'],
    'F': ['C'],
    'G': ['D'],
    'H': ['E']
}

print(bfs_graph(graph, 'A'))
```

### 图 DFS（深度优先搜索）

```python
def dfs_graph(graph, start):
    result = []
    visited = set()
    stack = [start]

    while stack:
        point = stack.pop()
        if point not in visited:
            visited.add(point)
            result.append(point)
            stack.extend(graph.get(point, []))

    return result


dfs_graph(graph, 'A')
```

### 图 DFS 后序（完成顺序）

```python
def dfs_postorder_graph(graph, start):
    result = []
    visited = set()

    def dfs(point):
        if point in visited:
            return

        visited.add(point)
        for next_point in graph.get(point, []):
            dfs(next_point)
        result.append(point)

    dfs(start)
    return result

dfs_postorder_graph(graph, 'A')
```

### 课程表（拓扑排序）

- [Course Schedule](https://leetcode.com/problems/course-schedule/description/)

```python
def can_finish(prerequisites):
    # 构建邻接表
    graph = {}
    for end, start in prerequisites:
        graph[start] = graph.get(start, set()) | {end}

    # 计算入度
    indegree = {}
    for start, ends in graph.items():
        indegree.setdefault(start, 0)
        for end in ends:
            indegree[end] = indegree.get(end, 0) + 1

    queue = {node for node, val in indegree.items() if val == 0}
    result = []
    while queue:
        node = queue.pop()
        result.append(node)
        for key in graph.get(node, []):
            indegree[key] -= 1
            if indegree[key] == 0:
                queue.add(key)

    return sum(indegree.values()) == 0, result, graph


can_finish([[1,4],[2,4],[3,1],[3,2]])
```

## Two Pointers（双指针）

### 两数之和

- 在升序数组中找到两个数，使其和等于目标值

```python
def two_sum(nums: list, target: int) -> list:
    slow = 0
    fast = len(nums) - 1

    while True:
        slow_fast_sum = nums[slow] + nums[fast]
        if slow_fast_sum == target:
            return [slow + 1, fast + 1]
        elif slow_fast_sum > target:
            fast -= 1
        else:
            slow += 1


numbers = [-1000,-1,0,1]
target = 1

two_sum(numbers, target)
```

## Dynamic Programming（动态规划）

### 最长公共子序列

```python
def LCS(X, Y):
    m = len(X) + 1
    n = len(Y) + 1
    martix = [[0] * n for _ in range(m)]

    for row in range(1, m):
        for col in range(1, n):
            if X[row - 1] == Y[col - 1]:
                martix[row][col] = martix[row - 1][col - 1] + 1
            else:
                martix[row][col] = max(martix[row][col - 1], martix[row - 1][col])

    return restore_seq(X, Y, m, n, martix)


def restore_seq(X, Y, m, n, martix):
    seq = []

    while m > 0 and n > 0:
        if X[m - 2] == Y[n - 2]:
            seq.append(X[m - 2])
            m -= 1
            n -= 1
        elif martix[m - 1][n - 2] < martix[m - 2][n - 1]:
            m -= 1
        else:
            n -= 1

    return "".join(seq)[::-1]


LCS("ABEFF", "ADEOFOOBP")
```

### 最长公共子串

```python
def LCS(X, Y):
    m = len(X)
    n = len(Y)
    martix = [[0] * (n + 1) for _ in range(m + 1)]
    max_x = 0
    max_y = 0
    seq = []

    for row in range(1, m + 1):
        for col in range(1, n + 1):
            if X[row - 1] == Y[col - 1]:
                martix[row][col] = martix[row - 1][col - 1] + 1
                if martix[row][col] >= martix[max_x][max_y]:
                    max_x, max_y = row, col

    while martix[max_x][max_y]:
        seq.append(X[max_x - 1])
        max_x -= 1
        max_y -= 1

    return "".join(seq)[::-1]


LCS("ABCBDCBABPLB", "BDCABPLL")
```

## Greedy Algorithm（贪心算法）

### 最少主持人数（区间调度）

```python
import heapq as host_heapq


def minimum_num_of_hosts(nums):
    nums = sorted(nums, key=lambda item: (item[0], item[1]), reverse=False)
    hosts = [-float("inf")]

    for start, end in nums:
        if hosts[0] <= start:
            host_heapq.heappop(hosts) 
        host_heapq.heappush(hosts, end)

    return len(hosts)


nums = [[547,612],[417,551],[132,132],
        [168,446],[95,747],[187,908],
        [115,712],[15,329],[612,900],
        [3,509],[181,200],[562,787],
        [136,268],[36,784],[533,573],
        [165,946],[343,442],[127,725],
        [557,991],[604,613],[633,721],
        [287,847],[414,480],[428,698],
        [437,616],[475,932],[652,886],
        [19,992],[132,543],[390,869],
        [754,903],[284,925],[511,951],
        [272,739]]
minimum_num_of_hosts(nums)
```

### 字符串全排列

```python
from itertools import permutations as string_permutations


def permute_string(string):
    return ["".join(item) for item in string_permutations(string)]


permute_string("ABCD")
```

```python
def permute_string(string):
    cur_idx = 0
    cur_substrs = [string]
    next_substrs = []

    while cur_idx < len(string) - 1:
        substr = cur_substrs.pop()

        for idx in range(cur_idx, len(string)):
            newstr = list(substr)
            newstr[idx], newstr[cur_idx] = newstr[cur_idx], newstr[idx]
            next_substrs.append("".join(newstr))

        if not cur_substrs:
            cur_substrs = next_substrs
            next_substrs = []
            cur_idx += 1

    return sorted(set(cur_substrs))


permute_string("ABCD")
```

### 不同路径（回溯搜索）

```python
def func(obstacle_grid):
    result = 0
    cur_grid = [(0,0)]

    while cur_grid:
        x, y = cur_grid.pop()
        row = len(obstacle_grid)
        col = len(obstacle_grid[x])

        if obstacle_grid[x][y]:
            continue
        if x == row - 1 and y == col - 1:
            result += 1
        if x + 1 < row:
            cur_grid.append((x + 1, y))
        if y + 1 < col:
            cur_grid.append((x, y + 1))

    return result


func([[0]])
```

## Sorting Algorithms（排序算法）

### 冒泡排序

```python
def sort(nums):
    n = len(nums)
    for i in range(n - 1):
        swapped = False

        for j in range(n - 1 - i):
            if nums[j] > nums[j + 1]:
                nums[j], nums[j + 1] = nums[j + 1], nums[j]
                swapped = True

        if not swapped:
            break

    return nums


sort([1,3,32,4,1,-100,23,17,23,6,-7,2,9,-1000])
```

### 选择排序

```python
def sort(nums):
    for i in range(len(nums) - 1):
        min_idx = i

        for j in range(i + 1, len(nums)):
            if nums[j] < nums[min_idx]:
                min_idx = j

        nums[i], nums[min_idx] = nums[min_idx], nums[i]

    return nums


sort([1,3,32,4,1,-100,23,17,23,6,-7,2,9,-1000])
```

### 插入排序

```python
def sort(nums):
    for i in range(1, len(nums)):
        base = nums[i]

        while i > 0 and base <= nums[i - 1]:
            nums[i] = nums[i - 1]
            i -= 1

        nums[i] = base

    return nums


sort([1,3,32,4,1,-100,23,17,23,6,-7,2,9,-1000])
```

### 快速排序

```python
def partition(nums, left, right):
    i, j = left, right

    while i < j:
        while i < j and nums[j] >= nums[left]:
            j -= 1
        while i < j and nums[i] <= nums[left]:
            i += 1
        nums[i], nums[j] = nums[j], nums[i]

    nums[left], nums[i] = nums[i], nums[left]
    return i


def sort(nums, left, right):
    if left >= right:
        return

    idx = partition(nums, left, right)
    sort(nums, left, idx - 1)
    sort(nums, idx + 1, right)


nums = [1,3,32,4,1,-100,23,17,23,6,-7,2,9,-1000]
left = 0
right = len(nums) - 1

sort(nums, left, right)
print(nums)
```

### 归并排序

```python
def merge(nums, left, mid, right):
    i, j = left, mid + 1
    tmp_array = []
    while i <= mid and j <= right:
        if nums[i] <= nums[j]:
            tmp_array.append(nums[i])
            i += 1
        else:
            tmp_array.append(nums[j])
            j += 1

    if i <= mid:
        tmp_array.extend(nums[i:mid + 1])
    else:
        tmp_array.extend(nums[j:right + 1])

    nums[left:right + 1] = tmp_array


def sort(nums, left, right):
    if left >= right:
        return

    mid = (right + left) // 2
    sort(nums, left, mid)
    sort(nums, mid + 1, right)
    merge(nums, left, mid, right)


nums = [1,3,32,4,1,-100,23,17,23,6,-7,2,9,-1000]
left = 0
right = len(nums) - 1

sort(nums, left, right)
print(nums)
```

### 堆排序

```python
def sift_down(nums, n, i):
    while True:
        max_idx = i
        l_idx = 2 * i + 1
        r_idx = 2 * i + 2

        if l_idx < n and nums[l_idx] > nums[max_idx]:
            max_idx = l_idx
        if r_idx <  n and nums[r_idx] > nums[max_idx]:
            max_idx = r_idx
        if max_idx == i:
            break

        nums[max_idx], nums[i] = nums[i], nums[max_idx]
        i = max_idx


def heap_sort(nums, n):
    for i in range(n // 2 - 1, -1, -1):
        sift_down(nums, n, i)
    for i in range(n - 1, -1, -1):
        nums[0], nums[i] = nums[i], nums[0]
        sift_down(nums, i, 0)
    return nums


nums = [1,3,32,4,1,-100,23,17,23,6,-7,2,9,-1000]
n = len(nums)

print(heap_sort(nums, n))
```

### 桶排序

```python
import random as bucket_random


def sort(nums):
    if len(nums) <= 1:
        return nums

    min_num = min(nums)
    max_num = max(nums)
    if min_num == max_num:
        return nums

    bucket_count = len(nums)
    bucket_size = (max_num - min_num) / bucket_count
    buckets = [[] for _ in range(bucket_count)]
    new_nums = []

    for num in nums:
        bucket_idx = int((num - min_num) / bucket_size)
        bucket_idx = min(bucket_idx, bucket_count - 1)
        buckets[bucket_idx].append(num)
    for bucket in buckets:
        bucket.sort()
        new_nums.extend(bucket)

    return new_nums


nums = [bucket_random.randrange(0, 100) / 100 for _ in range(10)]
print(nums)
print(sort(nums))
```

### 计数排序

```python
import random as counting_random


def sort(nums):
    if not nums:
        return []

    min_num = min(nums)
    max_num = max(nums)
    cnt_array = [0] * (max_num - min_num + 1)
    new_nums = []

    for num in nums:
        cnt_array[num - min_num] += 1

    for idx, cnt in enumerate(cnt_array):
        new_nums.extend([idx + min_num] * cnt)

    return new_nums


nums = [counting_random.randrange(0, 100) for _ in range(10)]
print(nums)
print(sort(nums))
```

### 基数排序

```python
import random as radix_random


def radix_sort_non_negative(nums):
    if not nums:
        return []

    max_num = max(nums)
    base = 1

    while base <= max_num:
        buckets = [[] for _ in range(10)]
        new_nums = []

        for num in nums:
            buckets[num // base % 10].append(num)
        for bucket in buckets:
            new_nums.extend(bucket)

        nums = new_nums
        base *= 10

    return nums


def sort(nums):
    negatives = [-num for num in nums if num < 0]
    positives = [num for num in nums if num >= 0]

    sorted_negatives = radix_sort_non_negative(negatives)
    sorted_positives = radix_sort_non_negative(positives)

    return [-num for num in reversed(sorted_negatives)] + sorted_positives


nums = [radix_random.randrange(0, 10000) for _ in range(8)]
print(nums)
print(sort(nums))
```

> 在输入满足特定约束时，非比较排序可以达到线性时间复杂度。

# 构造与资源管理

`__init__()` 在实例创建后初始化对象状态。`__new__()` 负责创建实例，大多数普通类只需要实现 `__init__()`。

## __init__

创建实例时，Python 会自动调用 `__init__()`。


```python
class Person:
    def __init__(self, name):
        self.name = name

    def print_name(self):
        print(self.name)


person = Person("yemengjie")
person.print_name()
```

## __new__

调用顺序是先执行 `__new__()`，再执行 `__init__()`。只有在控制实例创建过程或继承不可变类型等特殊场景中，才需要自定义 `__new__()`。


```python
class Person:
    def __new__(cls, name: str):
        print("call __new__")
        return super().__new__(cls)

    def __init__(self, name: str) -> None:
        print("call __init__")
        self.name = name


person = Person("yemengjie")
print(person.name)
```

## 上下文管理器

`__del__()` 调用时机不确定，不适合管理关键资源。文件、网络连接、锁等资源更推荐用 `with` 和上下文管理协议管理。


```python
class ManagedResource:
    def __enter__(self):
        print("open resource")
        return self

    def __exit__(self, exc_type, exc, traceback):
        print("close resource")
        return False

    def use(self):
        print("use resource")


with ManagedResource() as resource:
    resource.use()
```

# Lambda 与 map

`lambda` 表达式用于创建匿名函数，适合短小、一次性的表达式。逻辑较复杂时应使用 `def`。

## lambda

`lambda` 的主体只能是一个表达式，不能包含多条语句或文档字符串。


```python
square = lambda x: x * x
adder = lambda x=1, y=2: x + y
print(square(2))
print(adder())
print(adder(x=3, y=4))
```

## map

`map()` 把函数惰性地应用到一个或多个可迭代对象上。需要列表时显式调用 `list()`；简单转换通常优先使用列表推导。


```python
def fahrenheit(temp):
    return (9 / 5) * temp + 32


def celsius(temp):
    return (5 / 9) * (temp - 32)


c_temps = [0, 22.5, 40, 100]
f_temps = list(map(fahrenheit, c_temps))
print(f_temps)
print(list(map(celsius, f_temps)))
print([(9 / 5) * temp + 32 for temp in c_temps])
```

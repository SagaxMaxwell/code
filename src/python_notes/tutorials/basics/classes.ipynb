# 类定义

类用 `class` 定义，类名通常使用大写开头的单数名。类实例化后，可以通过点号访问属性和方法。

## 定义类

实例方法的第一个参数按照惯例命名为 `self`，表示当前实例。调用实例方法时不需要手动传入 `self`。


```python
class Person:
    def set_name(self, name):
        self.name = name

    def print_name(self):
        print(self.name)


person = Person()
person.set_name("yemengjie")
person.print_name()
print(person.name)
```

## 属性引用和实例化

类对象支持属性引用和实例化。对象可以包含任意数量和类型的数据。

# 常用内置函数

这个主题整理 `enumerate()`、`map()`、`filter()`、`reduce()`、`zip()`、`all()`、`any()` 和 `sorted()`。现代 Python 中，简单转换和过滤通常优先使用推导式；已有函数需要组合时再使用这些工具。

## enumerate

`enumerate()` 同时返回索引和值，常用于循环中需要位置的场景。


```python
items = ["one", "two", "three"]
for index, item in enumerate(items, 1):
    print(index, item)
```

## map、filter 和 reduce

`map()` 和 `filter()` 返回惰性迭代器。`reduce()` 会反复应用二元函数并返回最终值；若能用 `sum()`、`max()`、`min()` 或清晰循环表达，通常优先选择这些写法。


```python
from functools import reduce
from operator import mul


numbers = [1, 2, 3, 4, 5]
print(list(map(lambda x: x**2, numbers)))
print(list(filter(lambda x: x % 2 == 1, numbers)))
print(reduce(mul, numbers, 1))
print([x**2 for x in numbers])
```

## zip

`zip()` 把多个可迭代对象按位置配对，长度以最短对象为准。配合 `dict()` 可以快速生成映射。


```python
names = ["Mukesh", "Roni", "Chari"]
ages = [24, 50, 18]
for index, (name, age) in enumerate(zip(names, ages)):
    print(index, name, age)

stocks = ["reliance", "infosys", "tcs"]
prices = [2175, 1127, 2750]
print(dict(zip(stocks, prices)))
```

## all 和 any

`all()` 判断所有元素是否为真，空可迭代对象返回 `True`。`any()` 判断是否存在至少一个真值。


```python
print(all([4, 5, 1]))
print(all([1, 0, 6]))
print(all([]))
print(any([0, False, 3]))
print(any([]))
```

## sorted

`sorted()` 接收任意可迭代对象并返回新列表。`key` 参数指定排序键，`reverse=True` 表示降序。`list.sort()` 会原地修改列表并返回 `None`。


```python
values = [2, 8, 1, 4, 6, 3, 7]
print(sorted(values))
print(sorted(values, reverse=True))
print(values)

words = ["cccc", "b", "dd", "aaa"]
print(sorted(words, key=len))
print(sorted([15, 4, 11, 7], key=lambda x: x % 7))
```

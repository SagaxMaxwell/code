# 面向对象概览

面向对象编程通过类和对象组织状态与行为。类负责抽象共同特征，对象是类的具体实例。

## 核心概念

- 类：描述具有相同属性和方法的一组对象。
- 对象：由类创建的数据结构实例。
- 实例化：根据类创建具体对象的过程。
- 属性：保存对象状态的数据。
- 方法：定义在类中的函数。
- 继承：子类复用父类属性和方法。
- 方法重写：子类重新实现父类方法。


## 学习路径

- `classes.ipynb`: 类定义、实例和 `self`。
- `init.ipynb`: 构造函数、`__new__()` 和上下文管理器。
- `methods.ipynb`: 实例属性、类属性、实例方法、类方法和静态方法。
- `special.ipynb`: 特殊方法、字符串表示和运算符重载。
- `inheritance.ipynb`: 单继承、多继承、组合、方法重写和 `super()`。

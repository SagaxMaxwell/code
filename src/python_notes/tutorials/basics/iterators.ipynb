# 生成器与迭代器

容器把元素组织在一起，迭代器按需返回下一个值，生成器是一种特殊迭代器。理解迭代协议有助于处理大数据序列和流式数据。

## 生成器表达式

列表推导式会一次性创建列表，生成器表达式会按需产生元素。


```python
lst = [x * x for x in range(5)]
generator = (x * x for x in range(5))
print(lst)
print(generator)
print(next(generator))
print(next(generator))
```

## 生成器函数

包含 `yield` 的函数会返回生成器对象。执行到 `yield` 时函数暂停，下一次迭代从暂停位置继续。


```python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b


print(list(fibonacci(10)))
```

## 可迭代对象和迭代器

实现 `__iter__()` 的对象可迭代。实现 `__iter__()` 和 `__next__()` 的对象是迭代器。生成器是迭代器，列表和字典是可迭代对象但不是迭代器。


```python
from collections.abc import Iterable, Iterator


values = [1, 2, 3]
iterator = iter(values)
print(isinstance(values, Iterable))
print(isinstance(values, Iterator))
print(isinstance(iterator, Iterator))
print(next(iterator))
```

# 函数参数

函数调用时传入的是实参，函数定义中的名称是形参。参数传递方式包括位置参数、关键字参数、默认值、可变位置参数和可变关键字参数。

## 位置、关键字和默认值

位置参数按顺序匹配。关键字参数通过参数名匹配。默认值可简化常见调用。


```python
def student_info(name="jasper", age=6):
    print(f"{name} is {age} years old")


student_info("jasper", 6)
student_info(age=10)
student_info(name="Jacobi", age=18)
```

## 传递列表

列表是可变对象。函数内部修改传入列表会影响调用方。如果要保留原列表，可以传入切片副本。


```python
def print_models(unprinted, completed):
    while unprinted:
        current = unprinted.pop()
        completed.append(current)


unprinted_designs = ["iphone case", "robot pendant", "dodecahedron"]
completed_models = []
print_models(unprinted_designs[:], completed_models)
print(unprinted_designs)
print(completed_models)
```

## 可变参数

`*args` 收集多余位置参数为元组，`**kwargs` 收集多余关键字参数为字典。调用函数时，`*` 可拆包序列，`**` 可拆包字典。


```python
def make_pizza(size, *toppings):
    print(f"Making a {size}-inch pizza")
    for topping in toppings:
        print("-", topping)


def build_profile(first, last, **user_info):
    profile = {"first_name": first, "last_name": last}
    profile.update(user_info)
    return profile


make_pizza(12, "mushrooms", "green peppers")
print(build_profile("albert", "einstein", location="princeton", field="physics"))
```

# 函数定义

函数把一组语句封装成可复用逻辑，可以接收参数，也可以返回结果。函数体以 `def` 开头，并通过缩进确定范围。

## 基本结构

函数第一行可以写文档字符串，用于说明用途、参数和返回值。


```python
def say_hello():
    """打印问候语。"""
    print("hello")


def greeting(name):
    """按名称打印问候语。"""
    print("hello, " + name)


def add(x, y):
    """返回两个数字之和。"""
    return x + y


say_hello()
greeting("John")
print(add(1, 2))
```

## 函数名也是对象

函数名可以赋给变量，也可以作为参数传给其他函数。理解这一点有助于学习 `lambda`、`map()` 和回调函数。


```python
def square(x):
    return x * x


f = square
print(f(3))
```

## 小结

函数名应清晰、简洁，并使用小写字母和下划线。重要函数应包含简短 docstring。复杂逻辑应拆成更小函数，便于测试和复用。

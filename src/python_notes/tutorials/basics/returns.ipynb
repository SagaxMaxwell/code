# 返回值

函数可以通过 `return` 把结果交给调用方。返回值可以是简单值，也可以是列表、字典等复杂结构。不写 `return` 时，函数返回 `None`。

## 简单返回值

返回值通常赋给变量，供后续逻辑继续使用。


```python
def get_formatted_name(first_name, last_name, middle_name=""):
    if middle_name:
        full_name = f"{first_name} {middle_name} {last_name}"
    else:
        full_name = f"{first_name} {last_name}"
    return full_name.title()


print(get_formatted_name("jimi", "hendrix"))
print(get_formatted_name("john", "hooker", "lee"))
```

## 返回字典

函数可以把多个字段组织成字典返回，便于调用方继续读取或扩展。


```python
def build_person(first_name, last_name, age=None):
    person = {"first": first_name, "last": last_name}
    if age is not None:
        person["age"] = age
    return person


print(build_person("jimi", "hendrix"))
print(build_person("jimi", "hendrix", age=27))
```

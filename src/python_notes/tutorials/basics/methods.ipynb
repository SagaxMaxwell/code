# 属性与方法

属性保存状态，方法描述行为。属性可分为实例属性和类属性，方法可分为实例方法、类方法和静态方法。

## 实例属性和类属性

实例属性属于某个对象，类属性由所有实例共享。


```python
class Person:
    species = "human"

    def __init__(self, name="no one"):
        self.name = name


p1 = Person()
p2 = Person("yemengjie")
print(p1.species, p1.name)
print(p2.species, p2.name)
```

## 实例方法、类方法和静态方法

实例方法需要实例状态，类方法通过 `cls` 访问类状态，静态方法不依赖类或实例状态。


```python
class Person:
    species = "human"

    def __init__(self, name="no one"):
        self.name = name

    def instance_method(self):
        print("instance method", self.name)

    @classmethod
    def class_method(cls):
        print("class method", cls.species)

    @staticmethod
    def static_method():
        print("static method")


person = Person()
person.instance_method()
Person.class_method()
Person.static_method()
```

## 公开、受保护和私有

一个下划线开头表示受保护成员约定，两个下划线开头表示私有成员。私有成员不应在类外直接访问。


```python
class Person:
    def __init__(self, name="no one", age=18, address="bnuz"):
        self.name = name
        self.__age = age
        self._address = address

    def show_profile(self):
        print(f"name: {self.name}, age: {self.__age}, address: {self._address}")


person = Person()
person.show_profile()
print(person.name)
print(person._address)
```

# 字符串

字符串是字符的有序集合，可以用单引号、双引号或三引号创建。字符串不可变，支持索引、切片、拼接、重复和常用方法。

## 创建和访问

索引从 `0` 开始，负索引从末尾向前数。切片形式为 `sequence[start:end:step]`，遵循左闭右开规则。


```python
text = "hello crash course"
print(text[6:11])
print(text[:-7])
print(text[::2])
print(text[::-1])
print(len(text))
```

## 转义和多行字符串

反斜杠用于转义。原始字符串前缀 `r` 可以让反斜杠保持字面含义。三引号可创建多行字符串。


```python
print("Ro\na\tldo")
print(r"Ro\na\tldo")

sample = """
This is sample,
it is very easy.
"""
print(sample)
```

## 格式化和方法

现代代码优先使用 f-string。常见方法包括 `title()`、`upper()`、`lower()`、`split()`、`join()`、`replace()`、`startswith()` 和 `strip()`。


```python
name = "jasper"
age = 6
print(f"My name is {name.title()}, and I am {age} years old.")

words = ["j", "a", "s", "p", "e", "r"]
print("".join(words))
print("a,b,c,d".split(","))
print("Python is good".replace("Python", "python"))
print("this is a book\n".strip())
```

## 类型转换

字符串和基本类型之间通常用构造函数转换，例如 `str()`、`int()`、`float()` 和 `bool()`。注意非空字符串转布尔值总是 `True`。


```python
print(str(100))
print(str(100.1))
print(str(True))
print(bool("False"))
print(bool(""))
print(int("123"))
print(float("123.4"))
```

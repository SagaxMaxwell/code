# itertools

`itertools` 是标准库中的迭代器工具模块，适合高效组合、排列和串联可迭代对象。

## permutations 和 combinations

`permutations()` 生成排列，顺序重要；`combinations()` 生成组合，顺序不重要。


```python
from itertools import combinations, permutations


print(list(permutations([1, 2, 3], 2)))
print(list(combinations([1, 2, 3, 4], 3)))
```

## product

`product()` 生成笛卡尔积，`repeat` 参数可以重复同一个输入序列。


```python
from itertools import product


print(list(product("ABCD", "xy")))
print(list(product([0, 1], repeat=3)))
print(list(product([1, 2, 3], [4, 5], [6, 7])))
```

## chain 和 accumulate

`chain()` 把多个可迭代对象串成一个序列。`accumulate()` 返回每一步的累计结果。


```python
from itertools import accumulate, chain
from operator import add, mul


print(list(chain(range(3), range(12, 15))))
print(list(accumulate([1, 3, 4, 10, 4], add)))
print(list(accumulate([1, 3, 4, 10, 4], mul)))
```

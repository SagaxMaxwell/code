# 进制转换

程序中常见进制包括二进制、八进制、十进制和十六进制。Python 可以直接书写不同进制的整数字面量，也可以用内置函数转换。

## 字面量

`0b` 表示二进制，`0o` 表示八进制，`0x` 表示十六进制。


```python
print(0b1101010)
print(0o125)
print(0x3f2a1)
```

## 函数转换

`int(value, base)` 可以按指定进制解析字符串。`bin()`、`oct()` 和 `hex()` 可以把整数转成对应进制字符串。


```python
print(int("1101010", 2))
print(int("125", 8))
print(int("3f2a1", 16))

print(bin(106))
print(oct(85))
print(hex(258721))
print(bin(int("125", 8)))
print(hex(int("1101010", 2)))
```

# 集合

集合 `set` 是无序、可迭代、可变且不包含重复元素的数据类型。它常用于去重、成员判断和数学集合运算。

## 创建

集合可以用花括号或 `set()` 创建。空花括号 `{}` 创建的是字典，空集合应使用 `set()`。


```python
print({1, 2, 3})
print({1.0, "Hello", (1, 2, 3)})
print(set("python"))
print(set([1, 2, 4, 4, 3, 3, 3]))
print(type({}))
print(type(set()))
```

## 修改和删除

`add()` 添加单个元素，`update()` 添加多个元素。`discard()` 删除不存在元素时不报错，`remove()` 删除不存在元素时会抛出 `KeyError`。


```python
my_set = set()
my_set.add(8)
my_set.update([10, 11])
print(my_set)

my_set.discard(8)
print(my_set)
my_set.clear()
print(my_set)
```

## 集合运算

集合支持并集、交集、差集和对称差集，对应运算符是 `|`、`&`、`-` 和 `^`。


```python
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}
print(A | B)
print(A & B)
print(A - B)
print(A ^ B)
```

## frozenset

`frozenset` 是不可变集合，可哈希，因此可以作为字典键或另一个集合的元素。


```python
A = frozenset([1, 2, 3, 4])
B = frozenset([3, 4, 5, 6])
print(A.isdisjoint(B))
print(A.intersection(B))
print(frozenset({1, 2, 3, 4, 5, 6}))
```

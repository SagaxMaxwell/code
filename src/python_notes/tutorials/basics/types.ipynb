# 变量与类型

Python 变量不需要提前声明。变量只是绑定到对象的名称，真正拥有类型的是对象本身。

## 变量

变量必须先赋值再使用。同一个变量名可以在不同时间绑定到不同类型的对象。


```python
counter = 100
weight = 76.5
name = "Blade"
print(counter, weight, name)

age = 100
print(age)
age = "some string"
print(age)
```

## 命名

变量名只能包含字母、数字和下划线，不能以数字开头，也不能使用 Python 关键字。命名应简洁且有描述性。


```python
a = b = c = 100
print(a, b, c)

a, b, c = 1, 2.0, "China"
print(a, b, c)
```

## 基本类型

常见标准类型包括数字、字符串、列表、元组、集合和字典。数字类型包括 `int`、`float`、`bool` 和 `complex`。


```python
a, b, c, d = 20, 5.5, True, 4 + 3j
print(type(a), type(b), type(c), type(d))
print(True + 1)
print(isinstance(a, int))
print(isinstance(True, float))
```

## 引用

变量保存的是对象引用。`id()` 可以查看对象标识。不可变对象发生“修改”时，通常会绑定到新对象；可变对象可以在原对象上改变内容。


```python
number = 3
print(hex(id(number)))
number = 5
print(hex(id(number)))

profile = {"name": "c++"}
print(hex(id(profile)))
profile["name"] = "python"
print(hex(id(profile)))
```

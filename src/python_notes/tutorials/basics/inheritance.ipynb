# 继承

继承让子类复用父类属性和方法，也可以在父类基础上扩展新行为。继承适合表达 `is-a` 关系；如果只是复用能力，组合往往更清晰。

## 单继承

子类可以显式调用父类构造函数，也可以用 `super()` 调用方法解析顺序中的下一个实现。


```python
class Person:
    def __init__(self, name, age, weight):
        self.name = name
        self.age = age
        self.__weight = weight

    def show_basic_info(self):
        print(f"{self.name}: age={self.age}, weight={self.__weight} kg")


class Student(Person):
    def __init__(self, name, age, weight, grade):
        super().__init__(name, age, weight)
        self.grade = grade

    def show_student_info(self):
        print(f"{self.name}: age={self.age}, grade={self.grade}")


student = Student("ken", 10, 30, 3)
student.show_basic_info()
student.show_student_info()
```

## 多继承和组合

多继承会增加方法解析顺序和初始化复杂度。实际项目中应谨慎使用，优先选择组合或更清晰的继承结构。


```python
class Speaker:
    def __init__(self, name, topic):
        self.name = name
        self.topic = topic

    def speak(self):
        print(f"I'm {self.name}, my topic is {self.topic}")


class Sample:
    def __init__(self, name, topic):
        self.speaker = Speaker(name, topic)


test = Sample("Tim", "Python")
test.speaker.speak()
print(Sample.mro())
```

## 方法重写

子类可以定义同名方法覆盖父类实现。`super()` 可用于调用父类或 MRO 中的下一个类的方法。


```python
class Parent:
    def my_method(self):
        print("调用父类方法")


class Child(Parent):
    def my_method(self):
        print("调用子类方法")
        super().my_method()


child = Child()
child.my_method()
```

## 子类构造函数

子类定义自己的 `__init__()` 后，父类 `__init__()` 不会自动执行。如果需要父类初始化逻辑，应在子类构造函数中调用 `super().__init__(...)`。


```python
class Father:
    def __init__(self, name):
        self.name = name
        print(f"father name: {self.name}")

    def get_name(self):
        return f"Father {self.name}"


class Son(Father):
    def __init__(self, name):
        super().__init__(name)
        print("Son init")

    def get_name(self):
        return f"Son {self.name}"


son = Son("morningstar")
print(son.get_name())
```

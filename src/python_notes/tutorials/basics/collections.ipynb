# collections

`collections` 是标准库中的容器工具模块。这个主题重点介绍 `Counter` 和 `defaultdict`。

## Counter

`Counter` 用于统计元素出现次数，结果像一个字典，键是元素，值是计数。


```python
from collections import Counter


letters = Counter("blue")
words = Counter(["red", "blue", "red", "green", "blue", "blue"])
print(letters)
print(words)
print(words.most_common(2))
```

## defaultdict

`defaultdict` 是 `dict` 的子类。访问缺失键时，它会调用工厂函数创建默认值，而不是抛出 `KeyError`。


```python
from collections import defaultdict


counts = defaultdict(int)
for word in ["red", "blue", "red"]:
    counts[word] += 1
print(dict(counts))

labels = defaultdict(lambda: "Default Value")
labels["a"] = 42
print(labels["a"])
print(labels["missing"])
```

## 嵌套结构

工厂函数可以递归返回 `defaultdict`，用于构造树形结构。


```python
import json
from collections import defaultdict


def tree():
    return defaultdict(tree)


root = tree()
root["Page"]["Python"]["defaultdict"]["Title"] = "Using defaultdict"
root["Page"]["Python"]["defaultdict"]["Subtitle"] = "Create a tree"
root["Page"]["Java"] = None
print(json.dumps(root, indent=4))
```

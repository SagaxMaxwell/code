# 字典

字典是键值对集合，用于存储映射关系。键必须是可哈希对象，值可以是任意对象。字典是可变对象。

## 创建和访问

字典可以用花括号、`dict()` 或字典推导式创建。


```python
my_dict = {"key1": "value1", "key2": "value2"}
print(my_dict)
print(my_dict["key2"])

pairs = dict([(1, "a"), (2, "b"), (3, "c")])
print(pairs)

squares = {x: x * x for x in range(20) if x % 2 == 1}
print(squares)
```

## 修改和删除

字典可以修改已有键的值，也可以通过赋值新增键。删除键可用 `del` 或 `pop()`，清空字典可用 `clear()`。


```python
profile = {"name": "Zara", "age": 7}
profile["age"] = 8
profile["class"] = "First"
print(profile)
print(profile.pop("class"))
print(profile)
```

## 遍历

`items()` 遍历键值对，`keys()` 遍历键，`values()` 遍历值。需要顺序时可以配合 `sorted()`。


```python
favorite_languages = {"jen": "python", "sarah": "c", "edward": "ruby", "phil": "python"}
for name, language in favorite_languages.items():
    print(f"{name}: {language}")

for name in sorted(favorite_languages.keys()):
    print(name.title())

for language in set(favorite_languages.values()):
    print(language.title())
```

## 常用方法

常用方法包括 `copy()`、`get()`、`items()`、`keys()`、`pop()`、`popitem()`、`update()` 和 `values()`。


```python
dict1 = {1: "Python", 2: "Java", 3: "Ruby", 4: "Scala"}
dict2 = dict1.copy()
print(dict2.get(1))
print(dict2.items())
print(dict2.keys())
dict2.pop(4)
dict2.update({3: "Scala"})
print(dict2.values())
```

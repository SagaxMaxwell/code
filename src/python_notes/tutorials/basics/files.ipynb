# 文件读写

Python 使用文件对象与外部文件交互。现代代码推荐使用 `with` 语句或 `pathlib.Path` 管理路径和文件生命周期。

## 文件类型和模式

文本文件以字符编码保存内容，二进制文件以字节保存内容。常见文本模式包括 `r`、`w`、`a`、`r+`、`a+` 和 `x`。


```python
from pathlib import Path


path = Path("workData.txt")
path.write_text("line 1\nline 2\nline 3\n", encoding="utf-8")
print(path.read_text(encoding="utf-8"))
```

## 打开和读取

`with path.open(...) as file_object` 会在代码块结束后自动关闭文件。读取方法包括 `read()`、`readline()`、`readlines()` 和直接遍历文件对象。


```python
with path.open("r", encoding="utf-8") as work_data:
    print(work_data.readline())
    print(work_data.readline())

with path.open("r", encoding="utf-8") as work_data:
    print(work_data.readlines())

with path.open("r", encoding="utf-8") as work_data:
    for line in work_data:
        print(line, end="")
```

## 写入和追加

`write_text()` 会整体写入文本。`open("a")` 会追加到文件末尾。写入非字符串数据前需要显式转换为字符串。


```python
with path.open("a", encoding="utf-8") as work_data:
    work_data.write("line 4\n")

values = [1234, 5678, 9012]
Path("workData1.txt").write_text("".join(f"{value}\n" for value in values), encoding="utf-8")
print(path.read_text(encoding="utf-8"))
```

## 编辑文件

修改文件中间内容的常见做法是读取原内容，生成新内容，再整体写回文件。直接使用 `w` 模式会清空原内容，使用前需要确认。


```python
lines = path.read_text(encoding="utf-8").splitlines(keepends=True)
lines.insert(1, "This goes between line 1 and 2\n")
path.write_text("".join(lines), encoding="utf-8")
print(path.read_text(encoding="utf-8"))
```

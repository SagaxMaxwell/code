# 打包与解包

`*` 和 `**` 可以在赋值、函数定义和函数调用中打包或解包数据。它们能让序列和字典更自然地传递给函数。

## 序列解包

星号变量会收集中间剩余元素，结果是列表。


```python
a, *b, c = [2, 7, 5, 6, 3, 4, 1]
print(a)
print(b)
print(c)
```

## 函数调用解包

调用函数时，`*args` 解包列表或元组，`**kwargs` 解包字典。


```python
def repeat(count, name):
    for _ in range(count):
        print(name)


args = [2, "cats"]
kwargs = {"count": 2, "name": "dogs"}
repeat(*args)
repeat(**kwargs)
```

## 函数定义打包

定义函数时，`*args` 收集位置参数，`**kwargs` 收集关键字参数。


```python
def show_arguments(*args, **kwargs):
    print("位置参数：", args)
    print("关键字参数：", kwargs)


options = {"foo": 42, "bar": 7}
show_arguments(*[3, 4, 9], **options)
show_arguments(3, 4, 9, foo=42, bar=7)
```

# 目录操作

目录操作包括创建目录、列出目录内容、筛选文件和子目录，以及递归遍历。现代 Python 推荐优先使用 `pathlib.Path`。

## 创建示例目录

下面创建一个小型目录树，便于演示遍历操作。


```python
from pathlib import Path


base = Path("my_directory")
for directory in [base / "sub", base / "logs", base / "config"]:
    directory.mkdir(parents=True, exist_ok=True)

for file_path in [base / "file1.py", base / "file2.csv", base / "file3.txt", base / "sub" / "bar.py", base / "sub" / "foo.py"]:
    file_path.parent.mkdir(parents=True, exist_ok=True)
    file_path.touch(exist_ok=True)

print(sorted(entry.name for entry in base.iterdir()))
```

## iterdir 和 scandir

`Path.iterdir()` 返回目录下的 `Path` 对象。`os.scandir()` 返回带文件类型信息的目录项，适合较底层扫描。


```python
import os


for entry in base.iterdir():
    print(entry.name)

with os.scandir(base) as entries:
    for entry in entries:
        print(entry.name, entry.is_file(), entry.is_dir())
```

## 列出文件和目录

`is_file()` 和 `is_dir()` 可用于筛选文件或目录。`rglob()` 支持递归匹配。


```python
print("files")
for entry in base.iterdir():
    if entry.is_file():
        print(entry.name)

print("dirs")
for entry in base.iterdir():
    if entry.is_dir():
        print(entry.name)

print("recursive")
for path in base.rglob("*.py"):
    print(path)
```

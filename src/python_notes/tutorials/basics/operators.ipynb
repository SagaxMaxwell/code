# 运算符

运算符用于完成算术、比较、赋值、位运算、逻辑判断、成员判断和身份判断。

## 算术运算符

常见算术运算符包括 `+`、`-`、`*`、`/`、`//`、`%` 和 `**`。`/` 返回浮点数，`//` 返回向下取整后的结果。


```python
a = 24
b = 5
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a // b)
print(a % b)
print(a**b)
print(-5 // 2)
print(-123 % 10)
```

## 比较和赋值

比较运算符返回布尔值。复合赋值如 `+=`、`*=` 会基于当前值计算后再赋回变量。


```python
a = 21
b = 10
print(a == b)
print(a != b)
print(a > b)

c = 0
c += a
c *= 2
c //= 3
print(c)
```

## 位运算符

位运算符会把整数按二进制位处理，包括 `&`、`|`、`^`、`~`、`<<` 和 `>>`。


```python
a = 60
b = 13
print(a & b)
print(a | b)
print(a ^ b)
print(~a)
print(a << 2)
print(a >> 2)
```

## 逻辑、成员和身份

`and`、`or`、`not` 用于逻辑判断；`in` 和 `not in` 用于成员判断；`is` 判断两个变量是否引用同一个对象，值相等应使用 `==`。


```python
items = [1, 2, 3, 4, 5]
print(2 in items)
print(10 not in items)
print(True and False)
print(True or False)

left = [1, 2, 3]
right = [1, 2, 3]
alias = left
print(left == right)
print(left is right)
print(left is alias)
```

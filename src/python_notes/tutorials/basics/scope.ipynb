# 作用域

变量的可见范围称为作用域。Python 名称查找遵循 LEGB 规则：Local、Enclosing、Global、Built-in。

## 局部和全局

函数会引入局部作用域。`if`、`for`、`while` 等普通代码块不会引入新作用域。


```python
message = "outer"
if True:
    message = "changed in block"
print(message)


def demo():
    inner_message = "inner"
    print(inner_message)


demo()
```

## global 和 nonlocal

内部作用域要修改全局变量时使用 `global`。内部函数要修改外层非全局作用域变量时使用 `nonlocal`。


```python
num = 1


def update_global():
    global num
    num = 123


update_global()
print(num)


def outer():
    value = 10

    def inner():
        nonlocal value
        value = 100

    inner()
    return value


print(outer())
```

## 避免覆盖内置名

不要把变量或函数命名为 `sum`、`list`、`dict` 等内置名称，否则会降低可读性，并可能遮蔽内置函数。

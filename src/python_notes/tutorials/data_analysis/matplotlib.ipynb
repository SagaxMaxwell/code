## 折线图


`matplotlib.pyplot.plot()` 可以绘制单条线，也可以在一次调用中绘制多条线。

画单条线的函数签名：

```python
matplotlib.pyplot.plot([x], y, [fmt], *, data=None, **kwargs)
```

画多条线的函数签名：

```python
matplotlib.pyplot.plot([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)
```

参数说明：

- `x`、`y`：点或线的节点数据，`x` 表示 `x` 轴数据，`y` 表示 `y` 轴数据。数据可以是列表或数组。
- `fmt`：可选参数，用于定义基础格式，例如颜色、标记和线条样式。
- `**kwargs`：可选参数，用于设置二维图形属性，例如标签、线宽和透明度。


对于图表来说，最简单的莫过于作出一个单一函数$y = f(x)$的图像。


````python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.default_rng(42)
````

在 Matplotlib 中，`Figure` 可以理解为容器，负责承载坐标轴、图像、文本和标签等对象。

创建一个 `Figure`：


````python
fig = plt.figure()
````

`plt.figure` 默认尺寸是 640 x 480，可以分别通过参数 `figsize` 和 `dpi` 修改尺寸和分辨率。还可以设置背景颜色之类的，详细内容参考文档。


````python
fig = plt.figure(figsize=(16, 9), dpi=200)
````

`Axes` 就是在 `Figure` 上面看到的图像，一个有边界的格子包括刻度和标签，最终还有画在上面的图表元素。


在当前 `Figure` 下创建一个 `Axes`


````python
ax = plt.axes()
````

`plt.axes` 默认是全屏的，也可以通过设置 `[left, bottom, width, height]` 设置位置和尺寸。


接下来就可以直接在上面绘制图形了。


````python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x))
````

用 `plt.subplots()` 可以同时创建 `Figure` 和 `Axes`。


````python
fig, ax = plt.subplots()
x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
````

可以使用 `pylab` 接口，也就是 `MATLAB` 风格接口绘制图形。


如果需要在同一幅图形中绘制多根线条，只需要多次调用 `plot` 函数即可：


````python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
````

> **注意**
>
> `plt.plot()` 和 `ax.plot()` 看上去效果类似，但它们代表两种不同的绘图方式。`plt.plot()` 是 Matplotlib 为兼容 MATLAB 而提供的状态式接口，`ax.plot()` 是更适合复杂图形的面向对象接口。


### 线条颜色和风格


通过指定 `color` 关键字参数可以调整颜色，这个字符串类型参数基本上能用来代表任何你能想到的颜色。可以通过多种方式指定颜色参数：
所有 `HTML` 颜色名称可以在[这里](https://www.w3schools.com/colors/colors_names.asp)找到。


````python
plt.plot(x, np.sin(x - 0), color='blue')  # 颜色名
plt.plot(x, np.sin(x - 1), color='g')  # 颜色简写
plt.plot(x, np.sin(x - 2), color='0.75')  # 灰阶
plt.plot(x, np.sin(x - 3), color='#FFDD44')  # 十六进制
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3))  # RGB 元组
plt.plot(x, np.sin(x - 5), color='chartreuse')  # HTML 颜色名
````

如果没有指定颜色，Matplotlib 会在一组默认颜色中循环绘制每一条线。

类似地，可以通过 `linestyle` 关键字参数指定线条风格：


````python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')

# 线型简写
plt.plot(x, x + 4, linestyle='-')  # 实线
plt.plot(x, x + 5, linestyle='--') # 虚线
plt.plot(x, x + 6, linestyle='-.') # 点划线
plt.plot(x, x + 7, linestyle=':');  # 点线
````

如果喜欢更简洁的代码，这些 linestyle 和 color 参数能够合并成一个非关键字参数，传递给 plt.`plot()` 函数：


````python
plt.plot(x, x + 0, '-g')  # 绿色实线
plt.plot(x, x + 1, '--c') # 天青色虚线
plt.plot(x, x + 2, '-.k') # 黑色长短点虚线
plt.plot(x, x + 3, ':r');  # 红色点线
````

上面的单字母颜色码是 RGB 颜色系统以及 CMYK 颜色系统的缩写，被广泛应用在数字化图像的颜色系统中。

还有很多其他的关键字参数可以对折线图的外观进行精细调整，可参考文档。


````python
x = np.linspace(0, 10, 20)
plt.plot(x, x + 0,
    color='red',
    linestyle='-',
    marker='*',
    markerfacecolor='white',
    markeredgecolor='gray',
    markeredgewidth=2,
    linewidth=0.5
    )
````

不同的 `marker` 风格。


````python
rng = np.random.default_rng(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.random(5), rng.random(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8)
````

### 坐标轴范围

Matplotlib 会自动选择合适的坐标轴范围来绘制图像，但有些场景需要手动调整。


````python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
````

如果某些情况下希望将坐标轴反向，可以通过上面的函数实现，将参数顺序颠倒即可：


````python
plt.plot(x, np.sin(x))

plt.xlim(10, 0)
plt.ylim(1.2, -1.2);
````

相关函数还有 `plt.axis()`。注意，它不是 `plt.axes()`；`plt.axis()` 可以通过一次函数调用设置 `x` 轴和 `y` 轴范围，只需要传入 `[xmin, xmax, ymin, ymax]` 列表。


````python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.axis([-1, 11, -1.5, 1.5])
````

当然 `plt.axis()` 函数不仅能设置范围，还能像下面代码一样将坐标轴调整到刚好足够绘制折线图像的大小：


````python
plt.plot(x, np.sin(x))
plt.axis('tight')
````

还可以通过设置 `'equal'` 参数设置 `x` 轴与 `y` 轴使用相同的长度单位：


````python
plt.plot(x, np.sin(x))
plt.axis('equal')
````

更多关于设置 `axis` 属性的内容请查阅 `plt.axis` 函数的文档。


### 折线图标签

标题和坐标轴标签是最基础的图表标注，Matplotlib 提供了函数来方便地设置它们：


````python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x", size = 100)
plt.ylabel("sin(x)", size=10)
plt.tick_params(labelsize = 20)
````

这些标签的位置、大小和风格可以通过函数参数设置。更多信息可以参阅 Matplotlib 在线文档和对应函数的文档字符串。


````python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve", fontdict=dict(fontsize=16, color='red', style='italic', family='Times New Roman'))
plt.xlabel("x")
plt.ylabel("sin(x)")
````

当一幅图中绘制多条折线时，图例可以让图表更清晰。通过 `plt.legend()` 可以快速创建图例。


````python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend(loc='upper left', title="trigonometric", frameon=True, framealpha=0.2)
````

上图可见，`plt.legend()` 函数绘制的图例线条与图中的折线无论风格和颜色都保持一致。查阅 `plt.legend` 文档字符串可以获得更多相关信息。


在面向对象接口中，与其逐个调用上面的方法来设置属性，更常见的使用 `ax.set()` 方法来一次性设置所有的属性：


````python
ax = plt.axes()
ax.plot(x, np.sin(x))
# 全局字体
plt.rcParams['font.size'] = '16'
plt.rcParams['font.family'] = 'Times New Roman'
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot')
ax.plot()
````

也可以单独设置。


````python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set_xlim(0, 10)
ax.set_ylim(-2, 2)
ax.set_xlabel('x', color='red', fontsize=20)
ax.set_ylabel('sin(x)', color='blue', fontsize=20)
ax.set_title('A Simple Plot', color='green', fontsize=25, fontfamily='Times New Roman')
ax.tick_params(labelsize=20)
````

### 坐标轴位置


改变坐标轴位置，可以通过修改 `Spine` 对象的属性。Spine 指绘图区四周的边界线。

使用 `Axes.spines` 更改 `Spine` 的属性。实现诸如删除顶部和右侧的 `Spine`，或者将左侧 `Spine` 移动原点等功能。

`Axes.spines` 属性是一个字典对象，通过 `right`，`left`，`top`，`bottom` 键可以访问各个方位的 `Spine`。


使用 `set_position()` 方法设置 `Spine` 位置，其参数为 `tuple` 类型 ，包含两个需要设置的值 `(position_type, amount)`。一个是位置类型：

- `'outward'`，将 `Spine` 向图形区域外移动；
- `'axes'`，将 `Spine` 放在指定的 `Axes` 坐标（0 到 1 之间）；
- `'data'`，将 `Spine` 放在指定的数据坐标位置。

另外还有两个特殊值：

- `center` 表示 `('axes', 0.5)`。
- `zero` 表示 `('data', 0.0)`。


````python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4])
ax.spines['right'].set_color(None)
ax.spines['top'].set_color(None)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.plot()
# plt.show()
````

````python
fig, ax = plt.subplots()
x = np.linspace(-np.pi, np.pi, 100)
plt.plot(x, np.sin(x))

# 坐标轴移到原点
ax.spines["left"].set_position(("data", 0))
ax.spines["bottom"].set_position(("data", 0))

ax.spines['left'].set(color='red', linewidth=3)
# 隐藏上右边框
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.plot(1, 0, ">b", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^r", transform=ax.get_xaxis_transform(), clip_on=False)
````

有些情况下，数据的尺度不同，可以设置多个 y 轴。


````python
fig, ax = plt.subplots(figsize=(10, 6))

ax_twin = ax.twinx() # 共享 x 轴

x = np.linspace(1, 10, 50)
height = np.linspace(100, 200, 50)
weight = np.linspace(30, 80, 50)

ax.plot(x, height, "b-")
ax_twin.plot(x, weight, "r--")

ax.set_ylim(90, 210)
ax_twin.set_ylim(30, 80)

ax.tick_params(axis='y', colors='r')
ax_twin.tick_params(axis='y', colors='b')

ax.set_ylabel("Height(cm)", color='red')
ax_twin.set_ylabel("Weight(kg)", color='blue')
````

### 刻度


Matplotlib 的默认刻度格式在多数常见场景下表现良好，但特殊场景中可能需要更细致的个性化设置。


````python
fig, ax = plt.subplots(figsize=(8, 4))

x = np.linspace(0, 2 * np.pi, 1000)
ax.plot(x, np.sin(x), lw=3, label='Sine')
ax.grid(True)
ax.set_xlim(0, 2 * np.pi)
ax.minorticks_on() # 次刻度
````

设置主要刻度线


````python
ax.set_xticks([0, np.pi/2, np.pi, 1.5*np.pi, 2*np.pi])
ax.set_yticks([-1, -0.5,  0, 0.5, 1])
fig
````

下面设置刻度显示标签，并使用 Matplotlib 的 LaTeX 支持展示数学符号和公式。


````python
ax.set_xticklabels(['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
ax.minorticks_off()
fig
````

对于刻度比较少的，可以用上面的方法设置，如果比较多刻度的话，可以通过 `MultipleLocator` 来设置刻度的步长。


````python
fig, ax = plt.subplots(figsize=(16, 10))
x = np.linspace(-3*np.pi, 3 * np.pi, 1000)
ax.plot(x, np.sin(x), lw=3, label='Sine')
ax.grid(True)
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi / 2)) # 每隔 pi/2 一个刻度
ax.yaxis.set_major_locator(plt.MultipleLocator(0.5)) # 每隔 0.5 一个刻度
````

要修改显示的标签，需要修改刻度的 `formatter`。在这种情况中，可以使用 `plt.FuncFormatter`，这个对象能够接受一个用户自定义的函数来提供对于刻度标签的精细控制。


````python
def format_func(value, tick_number):
    N = int(np.round(2 * value / np.pi))
    if N == 0:
        return "0" # 0点
    elif N == 1:
        return r"$\frac{\pi}{2}$" # pi/2
    elif N == 2:
        return r"$\pi$" # pi
    elif N % 2 > 0:
        return r"$\frac{{%d}\pi}{2}$" %N # n*pi/2 n是奇数
    else:
        return r"${0}\pi$".format(N // 2) # n*pi n是整数

ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
fig
````

`plt.FuncFomatter()` 提供了对于图表刻度最高级的自定义和精细控制，并且当你需要创建需要印刷或出版的图表时非常方便。

## 子图

在数据分析中，并列展示不同图表有助于比较。Matplotlib 提供了子图概念：一个图表可以包含一组小的 `axes`，分别展示多个子图。

创建子图有很多方法，最常用的是 `plt.subplots()`。


下面来创建一个 2 x 3 网格的子图表，其中每一行的子图表共享它们的 `y` 轴，而每一列的子图表共享它们的 `x` 轴：


````python
fig, ax = plt.subplots(2, 3, sharex='all', sharey='all')
````

注意上面我们设置了 `sharex` 和 `sharey` 之后，内部子图表的 `x` 轴和 `y` 轴的标签就自动被去掉了。返回值中 `ax` 是一个 `NumPy` 数组，里面含有每一个子图表的实例，你可以使用 `NumPy` 索引的语法很简单的获得它们。


````python
# axes 是 2x3 数组
import itertools
for i, j in itertools.product(range(2), range(3)):
    ax[i, j].text(0.5, 0.5, str((i, j)), fontsize=18, ha='center')
fig
````

用 `plt.subplot_mosaic` 也可以创建不同大小的子图。


````python
def annotate_axes(ax, text, fontsize=18):
    ax.text(0.5, 0.5, text, transform=ax.transAxes,
            ha="center", va="center", fontsize=fontsize, color="darkgrey")
fig, ax = plt.subplot_mosaic([['upper left', 'right'],
                               ['lower left', 'right']],
                              figsize=(5.5, 3.5), layout="constrained")
for k in ax:
    ax[k].text(0.5, 0.5, f'ax["{k}"]', transform=ax[k].transAxes,
            ha="center", va="center", fontsize=18, color="darkgrey")
fig.suptitle('plt.subplot_mosaic()')
````

控制子图的间距


````python
fig, ax = plt.subplots(2, 2)
x = np.linspace(-np.pi, np.pi, 100)
ax[0, 0].plot(x, np.sin(x))
ax[0, 0].set_title('sin(x)')
ax[0, 1].plot(x, np.cos(x))
ax[0, 1].set_title('cos(x)')
ax[1, 0].plot(x, np.sin(x-1))
ax[1, 0].set_title('sin(x-1)')
ax[1, 1].plot(x, np.cos(x-1))
ax[1, 1].set_title('cos(x-1)')
plt.tight_layout(pad=2, w_pad=1, h_pad=2) # 控制子图之间的间距
````

## 散点图


`matplotlib.pyplot.scatter()` 用于绘制散点图。

函数签名：

```python
matplotlib.pyplot.scatter(
    x,
    y,
    s=None,
    c=None,
    marker=None,
    cmap=None,
    norm=None,
    vmin=None,
    vmax=None,
    alpha=None,
    linewidths=None,
    *,
    edgecolors=None,
    plotnonfinite=False,
    data=None,
    **kwargs,
)
```

参数说明：

- `x`、`y`：长度相同的数组，表示散点图中每个点的坐标。
- `s`：点的大小，默认值为 `20`，也可以传入数组为每个点分别设置大小。
- `c`：点的颜色，默认蓝色 `'b'`，也可以传入 `RGB` 或 `RGBA` 数组。
- `marker`：点的样式，默认值为 `'o'`。
- `cmap`：颜色映射，默认值为 `None`。当 `c` 是浮点数组时才使用。
- `norm`：亮度归一化方式，默认值为 `None`。当 `c` 是浮点数组时才使用。
- `vmin`、`vmax`：亮度范围设置。设置 `norm` 时会忽略这两个参数。
- `alpha`：透明度，范围通常是 `0` 到 `1`。默认值为 `None`。
- `linewidths`：标记边缘线宽。
- `edgecolors`：边缘颜色或颜色序列，常见值包括 `'face'` 和 `'none'`。
- `plotnonfinite`：是否绘制非有限值，例如 `inf`、`-inf` 或 `nan`。
- `**kwargs`：其他可选样式参数。


绘制散点图的方法是使用 `plt.scatter` 函数，它的使用方法和 `plt.plot` 类似。


````python
x = np.linspace(0, 10, 50)
y = np.sin(x)
plt.scatter(x, y, marker='o')
````

`plt.scatter` 和 `plt.plot` 的主要区别在于，`plt.scatter` 可以针对每个点设置不同属性（大小、填充颜色、边缘颜色等），还可以通过数据集合对这些属性进行设置。


````python
rng = np.random.default_rng(0)
x = rng.standard_normal(100)
y = rng.standard_normal(100)

colors = rng.random(100)
sizes = 1000 * rng.random(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar();  # 显示颜色对比条
````

注意图表右边有一个颜色对比条（这里通过 `colormap()` 函数输出），图表中的点大小的单位是像素。使用这种方法，散点的颜色和大小都能用来展示数据信息，在希望展示多个维度数据集合的情况下很直观。

例如，当我们使用 `Scikit-learn` 中的鸢尾花数据集，里面的每个样本都是三种鸢尾花中的其中一种，并带有仔细测量的花瓣和花萼的尺寸数据：


````python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

ax = plt.axes()

scatter = ax.scatter(features[0], features[1], alpha=0.5,
            s=100*features[3], c=iris.target, cmap='viridis')

ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[1])
ax.axis([4, 9, 1.5, 6])
ax.set_xticks([])
ax.set_yticklabels([])
legend1 = ax.legend(*scatter.legend_elements(), loc="upper left", title="Classes")
ax.add_artist(legend1)

handles, labels = scatter.legend_elements(prop="sizes", alpha=0.6)
legend2 = ax.legend(handles, labels, loc="upper right", title="Sizes", fontsize=12)
````

我们可以从上图中看出，可以通过散点图同时展示该数据集的四个不同维度：图中的 `(x, y)` 位置代表每个样本的花萼的长度和宽度，散点的大小代表每个样本的花瓣的宽度，而散点的颜色代表一种特定的鸢尾花类型。如上图的多种颜色和多种属性的散点图对于分析和展示数据集时都非常有帮助。


***
**plot 和 scatter 对比：性能提醒**
除了上面说的 `plt.plot` 和 `plt.scatter` 对于每个散点不同属性的支持不同之外，对于小的数据集来说，两者并无差别。当数据集增长到几千个点时，`plt.plot` 会明显比 `plt.scatter` 的性能要高。造成这个差异的原因是·`plt.scatter`·支持每个点使用不同的大小和颜色，因此渲染每个点时需要完成更多额外的工作。而·`plt.plot`·来说，每个点都是简单的复制另一个点产生，因此对于整个数据集来说，确定每个点的展示属性的工作仅需要进行一次即可。对于很大的数据集来说，这个差异会导致两者性能的巨大区别，因此，对于大数据集应该优先使用 `plt.plot` 函数。

## 误差可视化


### 基础误差条

调用 Matplotlib 函数即可创建基础误差条：


````python
x = np.linspace(0, 10, 30)
error = 0.1 * rng.standard_normal(30)
y = np.sin(x)
plt.errorbar(x, y, yerr=np.abs(error), fmt='-o', ecolor='lightblue', color='green', elinewidth=3, capsize=4)
````

除了上面介绍的参数，你还可以指定水平方向的误差条（xerr），单边误差条和其他很多的参数。参阅 `plt.errorbar` 的帮助文档获得更多信息。


### 连续误差

某些场景需要展示连续值的误差范围。虽然 Matplotlib 没有内建函数直接完成这个任务，但可以组合 `plt.plot()` 和 `plt.fill_between()` 实现。

这里采用简单的高斯过程回归方法，Scikit-Learn 提供了对应 API。这个方法适合在非参数化函数中获得连续误差。


````python
from sklearn.gaussian_process import GaussianProcessRegressor

# 定义模型和一些符合模型的点
model = lambda x: x * np.sin(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)

# 高斯过程拟合
gp = GaussianProcessRegressor()
gp.fit(xdata[:, np.newaxis], ydata)

xfit = np.linspace(0, 10, 1000)
yfit, std = gp.predict(xfit[:, np.newaxis], return_std=True)
dyfit = 2 * std  # 约 95% 区间
````

们现在有了 xfit、yfit 和 dyfit，作为对我们数据的连续拟合值以及误差限。当然我们也可以像上面一样使用 plt.errorbar 绘制误差条，但是事实上我们不希望在图标上绘制 1000 个点的误差条。于是我们可以使用 plt.fill_between 函数在误差限区域内填充一道浅色的误差带来展示连续误差：


````python
# 可视化结果
plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit, '-', color='gray')

plt.fill_between(x=xfit, y1=yfit - dyfit, y2=yfit + dyfit,
                 color='gray', alpha=0.2)
plt.xlim(0, 10);
````

注意上面我们调用 `fill_between` 函数：我们传递了的参数包括 `x` 值，`y` 值的低限，然后是 `y` 值的高限，结果是图表中介于低限和高限之间的区域会被填充。

上图为我们提供了一个非常直观的高斯过程回归展示：在观测点的附近，模型会被限制在一个很小的区域内，反映了这些数据的误差比较小。在远离观测点的区域，模型开始发散，反映了这时的数据误差比较大。


## 轮廓图

有些场景需要在二维图表中用轮廓线或颜色区域展示三维数据，例如等高线地图或温度分布图。Matplotlib 提供了三个常用函数：`plt.contour()` 绘制轮廓图，`plt.contourf()` 绘制填充轮廓图，`plt.imshow()` 展示图像。


首先使用一个简单的函数$z=f(x,y)$绘制一个轮廓图来进行说明，我们用来作为数组广播运算的例子：


````python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
````

轮廓图可以使用 `plt.contour` 函数进行创建。它接收三个参数：`x` 参数代表三维网格的平面横轴坐标，`y` 参数代表三维网格的平面纵轴坐标，而 `z` 参数代表三维网格的高度坐标。最容易用来准备这种网格数据的是 `np.meshgrid` 函数，可以将两个一维的数组构造成一个二维的网格。


````python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
````

下面我们可以绘制标准的轮廓线图表：


````python
fig, ax = plt.subplots(figsize=(10, 6))
ax.contour(X, Y, Z, colors='black')
````

图中值得注意的是，当使用单色绘制轮廓图时，虚线代表的是负数的数值，而实线代表的是正数。而轮廓线可以通过指定 cmap 参数来设置线条的色图。下例中展示了使用色图且绘制了更多的轮廓线的例子，会在整个数据范围区域内等距分布有 20 条轮廓线：


````python
fig, ax = plt.subplots(figsize=(10, 6))
contour = ax.contour(X, Y, Z, levels=8, cmap='jet')
ax.clabel(contour, fontsize=10, colors='k') # 显示等高线的值
````

使用 `plt.contourf()` 函数可以将上面的图改为填充轮廓图来解决这个问题，（注意函数名最后有个 f，代表填充 fill），这个函数的语法基本上与 `plt.contour()` 保持一致。

加上 `plt.colorbar()` 函数，在图表边上创建一个颜色图例用以展示颜色所表示的数值区域。


````python
fig, ax = plt.subplots(figsize=(10, 6))
contour = ax.contourf(X, Y, Z, levels=20, cmap='jet')
plt.colorbar(contour)
````

有了图例，很容易可以看出黑色区域代表着“峰”，而红色区域代表这“谷”。


`plt.imshow()` 可以将高度值转换为图像。


````python
def compute_mandelbrot(max_iter, threshold, nx, ny):
    # c 网格
    x = np.linspace(-2, 1, nx)
    y = np.linspace(-1.5, 1.5, ny)

    c = x[:,None] + 1j*y[None, :]
    z = np.zeros_like(c)
    # Mandelbrot 迭代
    mandelbrot_set = np.zeros((nx, ny))
    for j in range(max_iter):
        z = z**2 + c
        mandelbrot_set[abs(z) > threshold] = j
        c[abs(z) > threshold] = 0+0j

    return mandelbrot_set

mandelbrot_set = compute_mandelbrot(50, 2., 1000, 1000)

plt.imshow(mandelbrot_set.T, extent=[-2, 1, -1.5, 1.5], cmap='jet')
plt.colorbar()
````

### 三维轮廓图
mplot3d 也包含着能够创建三维浮雕图像的工具。就像二维的 `ax.contour` 图表，`ax.contour3D` 要求输入数据的格式是二维普通网格上计算得到的 `Z` 轴的数据值。下面我们展示一个三维的正弦函数轮廓图：


````python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
````

### 框线图和表面图

````python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color='black')
ax.set_title('wireframe');
````

注意虽然每个颜色填充的表面都是二维的，但是表面的边缘不需要是直线构成的。下面的例子使用 surface3D 绘制了一个部分极坐标网格，能够让我们切入到函数内部观察效果：

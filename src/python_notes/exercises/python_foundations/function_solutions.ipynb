# 函数练习参考答案

**学习目标：** 本参考答案按匿名函数、高阶函数、序列组合和聚合问题组织，强调用函数表达清晰的数据处理流程。

## 匿名函数与高阶函数

### 第 1 题：求给定区间内的偶数


```python
start, end = 4, 30
print([number for number in range(start, end) if number % 2 == 0])
```

### 第 2 题：字符串长度映射


```python
words = ["hi", "Paris", "Hamster", "New England"]
print(list(map(len, words)))
```

### 第 3 题：生成斐波那契数列


```python
def fibonacci(n):
    a, b = 0, 1
    result = []
    for _ in range(n):
        result.append(a)
        a, b = b, a + b
    return result


print(fibonacci(5))
```

### 第 4 题：统计奇数个数


```python
numbers = [1, 2, 3, 2, 3, 1, 3]
print(sum(number % 2 == 1 for number in numbers))
```

### 第 5 题：筛选大于给定值的元素


```python
scores = [66, 90, 68, 59, 76, 60, 88, 74, 81, 65, 92, 85]
threshold = 80
print([score for score in scores if score > threshold])
```

### 第 6 题：字符串列表转大写


```python
words = ["fox", "jump", "over", "dog"]
print([word.upper() for word in words])
```

## 序列组合与聚合

### 第 7 题：合并两个列表


```python
letters = ["a", "b", "c", "d", "e"]
numbers = [1, 2, 3, 4, 5]
print(list(zip(letters, numbers)))
```

### 第 8 题：求列表元素之和


```python
numbers = [3, 4, 6, 9, 34, 12]
print(sum(numbers))
```

### 第 9 题：矩阵转置


```python
matrix = [[1, 2], [3, 4], [5, 6], [7, 8]]
transposed = [list(row) for row in zip(*matrix)]
print(transposed)
```

### 第 10 题：求两个列表的共有元素


```python
list1 = [1, 2, 3, 5, 7, 9]
list2 = [2, 3, 5, 6, 7, 8]
print([number for number in list1 if number in set(list2)])
```

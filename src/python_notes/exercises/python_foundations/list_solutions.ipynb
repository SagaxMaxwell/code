# 列表练习参考答案

**学习目标：** 本参考答案按基础列表操作、重复值处理和嵌套列表问题组织，便于对照检查解题思路。

## 基础列表操作

### 第 1 题：交换首尾元素


```python
numbers = [11, 35, 9, 56, 22]
numbers[0], numbers[-1] = numbers[-1], numbers[0]
print(numbers)
```

### 第 2 题：统计元素出现次数


```python
numbers = [15, 2, 7, 10, 12, 3, 10, 28, 10]
target = 10
print(numbers.count(target))
```

### 第 3 题：求每个数字的各位数字之和


```python
def sum_digits(number):
    return sum(int(digit) for digit in str(abs(number)))


numbers = [14, 67, 18, 34]
print([sum_digits(number) for number in numbers])
```

### 第 4 题：输出第二大的元素


```python
numbers = [60, 11, 10, 4, 100]
unique_numbers = sorted(set(numbers), reverse=True)
print(unique_numbers[1])
```

### 第 5 题：输出偶数


```python
numbers = [2, 3, 5, 62, 14]
print([number for number in numbers if number % 2 == 0])
```

### 第 6 题：统计偶数和奇数个数


```python
numbers = [4, 7, 9, 64, 14]
even_count = sum(number % 2 == 0 for number in numbers)
odd_count = len(numbers) - even_count
print(f"Even = {even_count}, odd = {odd_count}")
```

## 重复值与集合运算

### 第 7 题：输出重复元素


```python
numbers = [10, 30, 20, 30, 40, 50, -20, 60, 60, -20]
duplicates = []
for number in numbers:
    if numbers.count(number) > 1 and number not in duplicates:
        duplicates.append(number)
print(duplicates)
```

### 第 8 题：移除空列表


```python
items = [5, 6, [], 3, [], 9]
print([item for item in items if item != []])
```

### 第 9 题：查找两个嵌套列表中的不同元素


```python
list1 = [[1, 2], [3, 4], [5, 6]]
list2 = [[3, 4], [5, 7], [1, 2]]
result = [item for item in list1 if item not in list2] + [item for item in list2 if item not in list1]
print(result)
```

### 第 10 题：在嵌套列表中随机获取一个元素


```python
import random

nested_numbers = [[4, 5, 5], [2, 7, 4], [8, 6, 3]]
row = random.choice(nested_numbers)
print(random.choice(row))
```

## 嵌套列表与频次统计

### 第 11 题：逆转每个子列表的元素顺序


```python
nested_numbers = [[4, 1, 6], [7, 8], [4, 10, 8]]
print([row[::-1] for row in nested_numbers])
```

### 第 12 题：输出没有重复过的元素


```python
numbers = [1, 2, 2, 5, 8, 4, 4, 8]
print([number for number in numbers if numbers.count(number) == 1])
```

### 第 13 题：获取出现频数大于等于给定值的元素


```python
numbers = [4, 6, 4, 3, 4, 4, 6, 6]
threshold = 3
result = []
for number in numbers:
    if numbers.count(number) >= threshold and number not in result:
        result.append(number)
print(result)
```

### 第 14 题：判断所有元素是否位于给定区间


```python
numbers = [4, 5, 6, 7, 3, 9]
lower, upper = 3, 10
print(all(lower <= number <= upper for number in numbers))
```

### 第 15 题：计算子串在字符串列表中出现的频次


```python
texts = ["一辈子", "好被子", "上辈子", "这杯子", "下辈子"]
keyword = "辈子"
print(sum(text.count(keyword) for text in texts))
```

### 第 16 题：按指定元素出现次数筛选元组


```python
tuples = [(4, 5, 5, 4), (5, 4, 3)]
target = 5
threshold = 2
print([item for item in tuples if item.count(target) >= threshold])
```

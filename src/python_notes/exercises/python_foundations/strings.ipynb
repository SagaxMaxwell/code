## 固定字符串处理

已知字符串 `text = "aAsmr3idd4bgs7Dlsf9eAF"`，完成下列任务。

1. 将字符串中的大写字母改为小写，小写字母改为大写。
2. 取出字符串中的数字，并组成一个新的字符串。
3. 统计每个字母出现的次数，忽略大小写。
4. 去除字符串中重复出现的字母，只保留第一次出现的字母。
5. 反转字符串。
6. 删除字符串中的数字。


```python
from collections import Counter

text = "aAsmr3idd4bgs7Dlsf9eAF"
```

### 任务 1：大小写互换


```python
print(text.swapcase())
```

### 任务 2：提取数字


```python
digits = "".join(char for char in text if char.isdigit())
print(digits)
```

### 任务 3：统计字母次数


```python
letter_counts = Counter(char.lower() for char in text if char.isalpha())
print(dict(letter_counts))
```

### 任务 4：去除重复字母


```python
seen = set()
unique_letters = []
for char in text:
    if char.isalpha() and char.lower() not in seen:
        unique_letters.append(char)
        seen.add(char.lower())

print("".join(unique_letters))
```

### 任务 5：反转字符串


```python
print(text[::-1])
```

### 任务 6：删除数字


```python
without_digits = "".join(char for char in text if not char.isdigit())
print(without_digits)
```

## 输入字符串处理

1. 接收一行字符，统计其中数字字符的个数。
2. 输入一个字符串，先提取下标为偶数的字符组成字符串 `A`，再提取下标为奇数的字符组成字符串 `B`，最后输出 `A + B`。
3. 将字符串中所有字符整体向后移动一位，即最后一个字符放到开头。
4. 判断输入字符串是否为回文。


### 任务 1：统计数字字符个数


```python
text = input("请输入字符串：")
digit_count = sum(char.isdigit() for char in text)
print(f"{digit_count} 个数字")
```

### 任务 2：按下标奇偶重组字符串


```python
text = input("请输入字符串：")
result = text[::2] + text[1::2]
print(result)
```

### 任务 3：循环右移一位


```python
text = input("请输入字符串：")
print(text[-1] + text[:-1] if text else "")
```

### 任务 4：判断回文


```python
text = input("请输入字符串：")
print(text == text[::-1])
```

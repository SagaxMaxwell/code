# 字典练习

**学习目标：** 本练习围绕字典创建、合并、分组、频次统计、嵌套访问和排序展开，适合巩固键值结构的实际用法。

## 字典创建与合并

### 第 1 题：求字典中所有值的和


```python
scores = {"a": 100, "b": 200, "c": 300}
print(sum(scores.values()))
```

### 第 2 题：合并两个字典


```python
dict1 = {"a": 10, "b": 8}
dict2 = {"d": 6, "c": 4}
print({**dict1, **dict2})
```

### 第 3 题：将矩阵转换为字典

字典的键为矩阵行号，从 `1` 开始。


```python
matrix = [[5, 6, 7], [8, 3, 2]]
print({index: row for index, row in enumerate(matrix, start=1)})
```

## 分组与频次统计

### 第 4 题：按元素分组


```python
numbers = [4, 6, 6, 4, 2, 2, 4, 8, 5, 8]
groups = {}
for number in numbers:
    groups.setdefault(number, []).append(number)
print(groups)
```

### 第 5 题：从第 `N` 个字符起排序字符串


```python
text = "paradox"
position = 3
print(text[: position - 1] + "".join(sorted(text[position - 1 :])))
```

### 第 6 题：统计列表词频


```python
numbers = [1, 1, 1, 5, 5, 3, 1, 3, 3, 1, 4, 4, 4, 2, 2, 2, 2]
frequency = {number: numbers.count(number) for number in sorted(set(numbers))}
print(frequency)
```

## 字典删除与嵌套访问

### 第 7 题：根据键删除字典项


```python
animals = {"fox": 21, "cat": 21, "dog": 22, "bat": 21}
animals.pop("dog", None)
print(animals)
```

### 第 8 题：提取嵌套字典中指定键对应的值


```python
data = {"fox": {"a": 7, "b": 9, "c": 12}, "dog": {"a": 15, "b": 19, "c": 20}}
key = "b"
print([inner[key] for inner in data.values() if key in inner])
```

### 第 9 题：将字典转换成嵌套列表


```python
data = {"fox": [1, 3, 4], "dog": [7, 6]}
print([[key, *value] for key, value in data.items()])
```

## 配对、切割与排序

### 第 10 题：将两个列表配对成字典


```python
values = ["coin", 10]
keys = ["name", "value"]
print(dict(zip(keys, values)))
```

### 第 11 题：按指定格式切割字符串并转换为字典


```python
text = "Jan = January; Feb = February; Mar = March;"
items = [item.strip() for item in text.split(";") if item.strip()]
print(dict(item.split(" = ") for item in items))
```

### 第 12 题：将元组列表转换为字典


```python
pairs = [("A", 1), ("B", 2), ("C", 3)]
print({key: [value] for key, value in pairs})
```

### 第 13 题：对字典中的键和值升序排序


```python
data = {"c": [3], "b": [12, 10], "a": [19, 4]}
print({key: sorted(data[key]) for key in sorted(data)})
```

### 第 14 题：对字典中的列表值求和


```python
data = {"fox": [6, 7, 4], "dog": [7, 6, 5]}
print({key: sum(values) for key, values in data.items()})
```

### 第 15 题：将字典中的两个列表配对


```python
data = {"month": [1, 2, 3], "name": ["Jan", "Feb", "March"]}
print(dict(zip(data["month"], data["name"])))
```

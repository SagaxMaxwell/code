**学习目标：** 本练习覆盖可变参数、返回值、默认参数、字符串处理和综合条件判断，重点训练函数拆分和复用。

## 可变参数与返回值

### 第 1 题：打印可变长度参数

编写函数 `func()`，接收任意数量的位置参数，并逐行打印每个参数。


```python
def func(*args):
    for value in args:
        print(value)


func(20, 40, 60)
func(80, 100)
```

### 第 2 题：同时返回加法和减法结果

编写函数 `calculation()`，输入两个数，并在一次返回中同时返回两数之和与两数之差。


```python
def calculation(left, right):
    return left + right, left - right


result = calculation(40, 10)
print(result)
```

## 默认参数与字符串处理

### 第 3 题：显示员工信息

编写函数 `show_employee()`，接收员工姓名和薪资。如果调用时缺少薪资参数，则默认显示 `9000`。


```python
def show_employee(name, salary=9000):
    print(f"Employee {name} salary is: {salary}")


show_employee("Ben", 9000)
show_employee("Ben")
```

### 第 4 题：排序连字符分隔的单词

编写函数 `sort_words()`，输入一个由连字符 `-` 分隔的单词序列，按字母顺序排序后再用连字符连接输出。


```python
def sort_words(text):
    words = text.split("-")
    return "-".join(sorted(words))


print(sort_words("green-red-yellow-black-white"))
```

## 综合条件判断

### 第 5 题：校验密码规则

编写程序，检查逗号分隔的密码序列，并输出符合条件的密码。规则如下：

- 至少包含一个小写字母 `[a-z]`；
- 至少包含一个大写字母 `[A-Z]`；
- 至少包含一个数字 `[0-9]`；
- 至少包含一个特殊字符 `$`、`#` 或 `@`；
- 密码长度为 `6` 到 `12` 个字符。


```python
def is_valid_password(password):
    has_lower = any(char.islower() for char in password)
    has_upper = any(char.isupper() for char in password)
    has_digit = any(char.isdigit() for char in password)
    has_symbol = any(char in "$#@" for char in password)
    has_valid_length = 6 <= len(password) <= 12
    return has_lower and has_upper and has_digit and has_symbol and has_valid_length


def filter_passwords(text):
    passwords = [password.strip() for password in text.split(",")]
    return [password for password in passwords if is_valid_password(password)]


sample = "ABd1234@1, a F1#,2w3E*,2We3345"
print(",".join(filter_passwords(sample)))
```

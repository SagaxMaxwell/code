## 区间筛选

### 第 1 题：按条件筛选区间数字

编写程序，输入两个整数 `a`、`b`，找出闭区间 `[a, b]` 中所有能被 `7` 整除但不是 `5` 的倍数的数，并用逗号连接输出。

> 样例输入：`1 100`
>
> 样例输出：`7,14,21,28,42,49,56,63,77,84,91,98`


```python
start, end = map(int, input("请输入两个整数，用空格分隔：").split())
result = [str(number) for number in range(start, end + 1) if number % 7 == 0 and number % 5 != 0]
print(",".join(result))
```

## 数值计算

### 第 2 题：计算阶乘

编写程序，计算给定整数 `n` 的阶乘，其中 `0 <= n <= 100`。

> 样例输入：`8`
>
> 样例输出：`40320`


```python
from math import factorial

n = int(input("请输入 n："))
print(factorial(n))
```

## 字典生成

### 第 3 题：生成平方字典

给定整数 `n`，生成一个字典，键为 `1` 到 `n` 的整数，值为对应整数的平方。

> 样例输入：`8`
>
> 样例输出：`{1: 1, 2: 4, 3: 9, 4: 16, 5: 25, 6: 36, 7: 49, 8: 64}`


```python
n = int(input("请输入 n："))
squares = {number: number ** 2 for number in range(1, n + 1)}
print(squares)
```

## 序列转换

### 第 4 题：生成列表和元组

从控制台接收一个逗号分隔的数字序列，并分别生成列表和元组。

> 样例输入：`34,67,55,33,12,98`
>
> 样例输出：
>
> `["34", "67", "55", "33", "12", "98"]`
>
> `("34", "67", "55", "33", "12", "98")`


```python
text = input("请输入逗号分隔的数字序列：")
items = text.split(",")
print(items)
print(tuple(items))
```

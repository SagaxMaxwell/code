```text
1    PythoN
2      JavA
3        C#
dtype: object
```


````python
sr = pd.Series(["php", "python", "java", "c#"])
sr.apply(lambda word: word[0].upper() + word[1:-1] + word[-1].upper() if len(word) > 1 else word.upper())
````

### 第 13 题：提取日期特征

从给定日期字符串 `Series` 中，输出月份中的第几天、一年中的第几天、`ISO` 周数和星期名称。

给定数据：

```python
date_series = pd.Series([
    "01 Jan 2015",
    "10-02-2016",
    "20180307",
    "2014/05/06",
    "2016-04-12",
    "2019-04-06T11:20",
])
```

期望输出：

```text
Day of month: [1, 2, 7, 6, 12, 6]
Day of year: [1, 276, 66, 126, 103, 96]
Week number: [1, 39, 10, 19, 15, 14]
Day of week: ["Thursday", "Sunday", "Wednesday", "Tuesday", "Tuesday", "Saturday"]
```


````python
date_series = pd.Series([
    "01 Jan 2015",
    "10-02-2016",
    "20180307",
    "2014/05/06",
    "2016-04-12",
    "2019-04-06T11:20",
])
dates = pd.to_datetime(date_series, format="mixed")

print("Day of month:", dates.dt.day.tolist())
print("Day of year:", dates.dt.dayofyear.tolist())
print("Week number:", dates.dt.isocalendar().week.tolist())
print("Day of week:", dates.dt.day_name().tolist())
````

### 第 14 题：筛选至少包含两个元音的单词

给定数据：

```python
color_series = pd.Series(["Red", "Green", "Orange", "Pink", "Yellow", "White"])
```

期望输出：

```text
1     Green
2    Orange
4    Yellow
5     White
dtype: object
```


````python
from collections import Counter


color_series = pd.Series(["Red", "Green", "Orange", "Pink", "Yellow", "White"])
vowels = set("aeiou")
mask = color_series.str.lower().apply(lambda word: sum(Counter(word)[vowel] for vowel in vowels) >= 2)
color_series[mask]
````

### 第 15 题：查找局部峰值位置

在给定 `Series` 中，找出左右两边都比它小的值的位置。

给定数据：

```python
nums = pd.Series([1, 8, 7, 5, 6, 5, 3, 4, 7, 1])
```

期望输出：

```python
[1, 4, 8]
```


````python
nums = pd.Series([1, 8, 7, 5, 6, 5, 3, 4, 7, 1])
peak_positions = np.where((nums.shift(1) < nums) & (nums.shift(-1) < nums))[0]
print(peak_positions.tolist())
````

### 第 16 题：计算 `Series` 自相关系数

给定公式：

$$
R_{yy}(l)=\sum_{n \in \mathbb{Z}} y(n)y(n-l)
$$

给定数据：

```python
num_series = pd.Series(np.arange(15) + np.random.normal(1, 10, 15))
```


````python
num_series = pd.Series(np.arange(15) + np.random.normal(1, 10, 15))
autocorrelations = [num_series.autocorr(i).round(2) for i in range(11)]
print(autocorrelations[1:])
````

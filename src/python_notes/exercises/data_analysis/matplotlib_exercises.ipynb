## 激活函数曲线

### 第 1 题：绘制常见激活函数

绘制 `sigmoid`、`tanh`、`ReLU`、`softplus` 和 `softmax` 函数图形。


```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

x = np.linspace(-6, 6, 100)

plt.figure(figsize= (6, 4), dpi= 100)
ax = plt.axes()

# 移动 left 和 bottom spines 到 (0, 0) 位置
ax.spines["left"].set_position(("data", 0))
ax.spines["bottom"].set_position('center')

# 隐藏 top 和 right spines.
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

#ax.set_xlim(-6, 6)
ax.set_ylim(-1.5, 1.5)

ax.plot(x, 1 / (1 + np.e**-x), linestyle = "-", label= "sigmod")
ax.plot(x, (np.e**x - np.e**-x) / (np.e**x + np.e**-x), linestyle = "--", label= "tanh")
ax.plot(x, list(map(lambda n: n if n > 0 else 0, x)), linestyle="-.", label="Relu")
ax.plot(x, np.log(1 + np.e**x), linestyle=":", label="softplus")
ax.legend(loc= "upper left")
```

### 第 2 题：绘制 `Swish` 函数

绘制 `swish(x) = x * sigmoid(beta * x)` 在 `beta = 0`、`0.5`、`1`、`100` 时的图形。

![`Swish` 函数示例](../../resources/images/swish.png)


```python
x = np.linspace(-6, 6, 100)

plt.figure(figsize= (6, 4), dpi= 100)
ax = plt.axes()

# 移动 left 和 bottom spines 到 (0, 0) 位置
ax.spines["left"].set_position(("data", 0))
ax.spines["bottom"].set_position(("data", 0))

# 隐藏 top 和 right spines.
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.axis([-4.5, 4.5, -2.5, 4.5])

ax.plot(x, x / (1 + np.e**(-x * 0)), linestyle="-", c = "g", label=r"$\beta = 0$")
ax.plot(x, x / (1 + np.e**(-x * 0.5)), linestyle="--", c = "c", label=r"$\beta = 0.5$")
ax.plot(x, x / (1 + np.e**(-x * 1)), linestyle="-.", c = "m", label=r"$\beta = 1$")
ax.plot(x, x / (1 + np.e**(-x * 100)), linestyle="-", c = "r", label=r"$\beta = 100$")
ax.legend()
```

## 练习 2：三角函数图形

### 第 3 题：绘制三角函数

![三角函数图形示例](../../resources/images/cosin.png)


```python
x = np.linspace(-6, 6, 100)

plt.figure(figsize=(6, 4), dpi=100, facecolor= 'w')
ax = plt.axes()

# 移动 left 和 bottom spines 到 (0, 0) 位置
ax.spines["left"].set_position(("data", 0))
ax.spines["bottom"].set_position(("data", 0))

# 隐藏 top 和 right spines.
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.set_xlim(-3.5, 3.5)
ax.set_yticks([-1, 1])
ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_xticklabels([r'-$\pi$', r'-$\frac{\pi}{2}$', 0, r'$\frac{\pi}{2}$', r'$\pi$'])

ax.plot(x, np.sin(x), linewidth= 2, c = "r", label= "sine")
ax.plot(x, np.cos(x), linewidth= 2, c = "b", label= "cosine")
ax.plot([np.pi*2/3]*2, [np.sqrt(3)/2, 0], linestyle = '--', c ='r')
ax.plot([np.pi*2/3]*2, [-0.5, 0], linestyle = '--', c ='b')
ax.scatter([np.pi*2/3]*2, [np.sqrt(3)/2, -0.5], c = 'r')

ax.annotate(r'$\sin(\frac{{2}{\pi}}{3}) = \frac{\sqrt{3}}{2}$', xy = (np.pi*2/3, np.sqrt(3)/2), xytext = (np.pi*2/3 + 1, np.sqrt(3)/2 + 0.2), arrowprops = {'color':'black', 'width':0.3, 'headwidth':5})
ax.annotate(r'$\cos(\frac{{2}{\pi}}{3}) = -\frac{1}{2}$', xy = (np.pi*2/3, -0.5), xytext = (np.pi*2/3 - 2,  -1), arrowprops=dict(arrowstyle="->", connectionstyle="angle3, angleA=0, angleB=-120"))

ax.legend(loc= "best")
```

## 柱状图

### 第 4 题：绘制人口变化柱状图

根据给定省份和人口变化数据绘制柱状图。


```python
import pandas as pd

numbers = [72, 60, 54.7, 28.1, 26, 18, 10, 8.12, 4.99, 3.5, 3.5, 1.07, 1, 1, 0, -0.4, -1, -1, -1.46, -2.83, -6, -10.02, -11, -13.6, -16, -22.49, -23.79, -25.6, -30.93, -46, -58]
provinces = ["浙江", "广东", "湖北", "江苏", "福建", "广西", "安徽", "海南", "山东", "宁夏", "重庆", "上海", "青海", "四川", "西藏","北京", "新疆", "陕西", "江西", "内蒙古", "贵州", "山西", "甘肃", "天津", "河北", "湖南", "吉林", "辽宁", "云南", "黒龙江", "河南"]

dic = dict(zip(provinces, [float(x) for x in numbers]))

dic_zero = {x:y for x, y in dic.items() if y == 0}
dic_down = {x:y for x, y in dic.items() if y < 0}
[dic.pop(x) for x in set(dic_down.keys()) | set(dic_zero.keys())]

# 字典转列表、画柱状图时会用到
dic = sorted(dic.items(), key = lambda x: x[1], reverse = False)
dic_down = sorted(dic_down.items(), key = lambda x: x[1], reverse = False)

plt.figure(figsize= (10, 14), facecolor= '#0f494a', dpi= 100)

# 解决中文显示问题
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size'] = '12'

ax = plt.axes()
ax.set_facecolor('#0f494a')
ax.set_xlim([-80, 80])
ax.set_yticks([])
ax.set_xticks([])

ax.spines["left"].set_edgecolor('w')
ax.spines["left"].set_position(("data", 0))
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_visible(False)

bar_up = [ax.barh(x, dic[x - 1][1], height = 0.3, facecolor='r') for x in range(1,  len(dic) + 1)]
bar_down = [ax.barh(x, dic_down[x][1], height = 0.3,  facecolor='g') for x in range(-len(dic_down), 0)]

ax.text(-9.4, -0.15, '0.0', c = 'w')
ax.text(3, -0.15, '、'.join([x for x in dic_zero]), c = 'w')

for u, v in zip(bar_up, dic):
    ax.text(- 10, u.patches[0].get_y() + u.patches[0].get_height() / 4, f"{v[0]}", c = 'w')
    ax.text(u.patches[0].get_width() + 3, u.patches[0].get_y() + u.patches[0].get_height() / 4, f"{v[1]}", c = 'w')

for d, v in zip(bar_down, dic_down):
    ax.text(3, d.patches[0].get_y() + d.patches[0].get_height() / 4, f"{v[0]}", c = 'w')
    ax.text(d.patches[0].get_width() - 10.5, d.patches[0].get_y() - d.patches[0].get_height() / 4, f"{v[1]}", c = 'w')

ax.set_title("内地各省市2021年年末常住人口变化情况\n单位: 万人", fontsize= 16, c = 'w')
```

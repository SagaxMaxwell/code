## 激活函数曲线

### 第 1 题：绘制常见激活函数

绘制 `sigmoid`、`tanh`、`ReLU`、`softplus` 和 `softmax` 函数图形。


```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(-6, 6, 200)


def sigmoid(values: np.ndarray) -> np.ndarray:
    return 1 / (1 + np.exp(-values))


def softmax(values: np.ndarray) -> np.ndarray:
    shifted = values - values.max()
    exp_values = np.exp(shifted)
    return exp_values / exp_values.sum()


fig, ax = plt.subplots(figsize=(6, 4), dpi=100)
ax.spines["left"].set_position(("data", 0))
ax.spines["bottom"].set_position("center")
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.set_ylim(-1.5, 1.5)

ax.plot(x, sigmoid(x), linestyle="-", label="sigmoid")
ax.plot(x, np.tanh(x), linestyle="--", label="tanh")
ax.plot(x, np.maximum(x, 0), linestyle="-.", label="ReLU")
ax.plot(x, np.logaddexp(0, x), linestyle=":", label="softplus")
ax.plot(x, softmax(x), linewidth=1, label="softmax over x")
ax.legend(loc="upper left")
fig.tight_layout()
```

### 第 2 题：绘制 `Swish` 函数

绘制 `swish(x) = x * sigmoid(beta * x)` 在 `beta = 0`、`0.5`、`1`、`100` 时的图形。

![`Swish` 函数示例](../../resources/images/swish.png)


```python
x = np.linspace(-6, 6, 200)


def sigmoid(values: np.ndarray) -> np.ndarray:
    return 1 / (1 + np.exp(-values))


fig, ax = plt.subplots(figsize=(6, 4), dpi=100)
ax.spines["left"].set_position(("data", 0))
ax.spines["bottom"].set_position(("data", 0))
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.axis([-4.5, 4.5, -2.5, 4.5])

for beta, linestyle, color in [
    (0, "-", "g"),
    (0.5, "--", "c"),
    (1, "-.", "m"),
    (100, "-", "r"),
]:
    y = x * sigmoid(beta * x)
    ax.plot(x, y, linestyle=linestyle, color=color, label=rf"$eta = {beta}$")

ax.legend()
fig.tight_layout()
```

## 练习 2：三角函数图形

### 第 3 题：绘制三角函数

![三角函数图形示例](../../resources/images/cosine.png)


```python
x = np.linspace(-6, 6, 200)
point_x = 2 * np.pi / 3

fig, ax = plt.subplots(figsize=(6, 4), dpi=100, facecolor="w")
ax.spines["left"].set_position(("data", 0))
ax.spines["bottom"].set_position(("data", 0))
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.set_xlim(-3.5, 3.5)
ax.set_yticks([-1, 1])
ax.set_xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
ax.set_xticklabels([r"-$\pi$", r"-$rac{\pi}{2}$", "0", r"$rac{\pi}{2}$", r"$\pi$"])

ax.plot(x, np.sin(x), linewidth=2, color="r", label="sine")
ax.plot(x, np.cos(x), linewidth=2, color="b", label="cosine")
ax.plot([point_x, point_x], [np.sqrt(3) / 2, 0], linestyle="--", color="r")
ax.plot([point_x, point_x], [-0.5, 0], linestyle="--", color="b")
ax.scatter([point_x, point_x], [np.sqrt(3) / 2, -0.5], color="r")

ax.annotate(
    r"$\sin(rac{2\pi}{3}) = rac{\sqrt{3}}{2}$",
    xy=(point_x, np.sqrt(3) / 2),
    xytext=(point_x + 0.7, np.sqrt(3) / 2 + 0.2),
    arrowprops={"arrowstyle": "->", "color": "black"},
)
ax.annotate(
    r"$\cos(rac{2\pi}{3}) = -rac{1}{2}$",
    xy=(point_x, -0.5),
    xytext=(point_x - 2.2, -1),
    arrowprops={"arrowstyle": "->", "connectionstyle": "angle3, angleA=0, angleB=-120"},
)

ax.legend(loc="best")
fig.tight_layout()
```

## 柱状图

### 第 4 题：绘制人口变化柱状图

根据给定省份和人口变化数据绘制柱状图。


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


numbers = [72, 60, 54.7, 28.1, 26, 18, 10, 8.12, 4.99, 3.5, 3.5, 1.07, 1, 1, 0, -0.4, -1, -1, -1.46, -2.83, -6, -10.02, -11, -13.6, -16, -22.49, -23.79, -25.6, -30.93, -46, -58]
provinces = ["浙江", "广东", "湖北", "江苏", "福建", "广西", "安徽", "海南", "山东", "宁夏", "重庆", "上海", "青海", "四川", "西藏", "北京", "新疆", "陕西", "江西", "内蒙古", "贵州", "山西", "甘肃", "天津", "河北", "湖南", "吉林", "辽宁", "云南", "黑龙江", "河南"]

changes = pd.DataFrame({"province": provinces, "change": numbers})
plot_data = changes.loc[changes["change"] != 0].sort_values("change")
zero_provinces = changes.loc[changes["change"] == 0, "province"].tolist()
colors = np.where(plot_data["change"] >= 0, "#e15759", "#59a14f")

plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False
plt.rcParams["font.size"] = 12

fig, ax = plt.subplots(figsize=(10, 14), facecolor="#0f494a", dpi=100)
ax.set_facecolor("#0f494a")
ax.barh(plot_data["province"], plot_data["change"], height=0.4, color=colors)
ax.axvline(0, color="white", linewidth=1)
ax.set_xlim(-80, 80)
ax.set_xlabel("变化人数（万人）", color="white")
ax.set_title("内地各省市 2021 年年末常住人口变化情况\n单位：万人", fontsize=16, color="white")
ax.tick_params(colors="white")

for spine in ax.spines.values():
    spine.set_visible(False)

for y, (_, row) in enumerate(plot_data.iterrows()):
    value = row["change"]
    label_x = value + 2 if value > 0 else value - 2
    ha = "left" if value > 0 else "right"
    ax.text(label_x, y, f"{value:g}", va="center", ha=ha, color="white")

if zero_provinces:
    ax.text(2, -0.8, f"零增长：{'、'.join(zero_provinces)}", color="white")

fig.tight_layout()
```

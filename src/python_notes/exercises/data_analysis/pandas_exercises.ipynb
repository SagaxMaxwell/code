## 练习 1：国家数据分析

给定国家、人口、面积和首都数据，完成下面的数据整理与筛选任务。


```python
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False

data = {
    "country": ["Belgium", "France", "Germany", "Netherlands", "United Kingdom"],
    "population": [11.3, 64.3, 81.3, 16.9, 64.9],
    "area": [30510, 671308, 357050, 41526, 244820],
    "capital": ["Brussels", "Paris", "Berlin", "Amsterdam", "London"],
}

df = pd.DataFrame(data)
df
```

### 任务 1：设置索引

将 `country` 列设置为索引。


```python
df = df.set_index("country")
df
```

### 任务 2：绘制人口和面积条形图

分别绘制各个国家的人口和面积条形图。面积以 `10000 km^2` 为单位展示。


```python
plot_df = pd.DataFrame({
    "population_million": df["population"],
    "area_10k_km2": df["area"] / 10000,
})

ax = plot_df.plot(kind="bar", figsize=(9, 5), rot=30)
ax.set_xlabel("country")
ax.set_ylabel("value")
ax.set_title("Population and Area by Country")
plt.tight_layout()
```

### 任务 3：筛选面积大于 `100000` 的国家


```python
df[df["area"] > 100000]
```

### 任务 4：添加人口密度列

添加 `density` 列，表示人口密度，计算公式为 `population * 1000000 / area`。


```python
df["density"] = df["population"] * 1_000_000 / df["area"]
df
```

### 任务 5：添加相对人口密度列

添加 `density_ratio` 列，表示每个国家人口密度与整体平均人口密度的比值。


```python
average_density = df["population"].sum() * 1_000_000 / df["area"].sum()
df["density_ratio"] = df["density"] / average_density
df
```

### 任务 6：筛选人口密度大于 `300` 的国家

输出这些国家的首都和人口。


```python
df.loc[df["density"] > 300, ["capital", "population"]]
```

### 任务 7：修改英国首都

把 `United Kingdom` 的首都改成 `Cambridge`。


```python
df.loc["United Kingdom", "capital"] = "Cambridge"
df
```

### 任务 8：筛选首都名称长度超过 `7` 的国家


```python
df[df["capital"].str.len() > 7]
```

### 任务 9：筛选首都名称包含 `am` 的国家


```python
df[df["capital"].str.contains("am", case=False, na=False)]
```

## 练习 2：Titanic 数据分析

读取 `resources/datasets/titanic.csv`，用 Pandas 完成下列任务。

- 计算不同性别的平均年龄。
- 计算所有乘客的平均存活率。
- 计算所有小于 `25` 岁乘客的存活率。
- 计算不同性别的存活率。
- 绘制不同客舱等级存活率的条形图，即 `Pclass` 列。


```python
titanic = pd.read_csv("../../resources/datasets/titanic.csv")
titanic.head()
```

### 任务 1：计算不同性别的平均年龄


```python
titanic.groupby("Sex")["Age"].mean()
```

### 任务 2：计算整体平均存活率


```python
titanic["Survived"].mean()
```

### 任务 3：计算小于 `25` 岁乘客的存活率


```python
titanic.loc[titanic["Age"] < 25, "Survived"].mean()
```

### 任务 4：计算不同性别的存活率


```python
titanic.groupby("Sex")["Survived"].mean()
```

### 任务 5：绘制不同客舱等级的存活率


```python
survival_by_class = titanic.groupby("Pclass")["Survived"].mean()
ax = survival_by_class.plot(kind="bar", figsize=(7, 4), rot=0)
ax.set_xlabel("Pclass")
ax.set_ylabel("Survival rate")
ax.set_ylim(0, 1)

for index, value in enumerate(survival_by_class):
    ax.text(index, value + 0.02, f"{value:.1%}", ha="center")

plt.tight_layout()
```

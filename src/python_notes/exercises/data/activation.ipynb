# 激活函数练习

绘制 `sigmoid`、`tanh`、`ReLU`、`softplus`、`softmax` 和 `Swish` 函数，训练曲线绘制、坐标轴设置和图例。

## 常见激活函数

在同一坐标系中绘制多条函数曲线。


```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(-6, 6, 200)


def sigmoid(values: np.ndarray) -> np.ndarray:
    return 1 / (1 + np.exp(-values))


def softmax(values: np.ndarray) -> np.ndarray:
    shifted = values - values.max()
    exp_values = np.exp(shifted)
    return exp_values / exp_values.sum()


fig, ax = plt.subplots(figsize=(6, 4), dpi=100)
ax.spines["left"].set_position(("data", 0))
ax.spines["bottom"].set_position("center")
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.set_ylim(-1.5, 1.5)
ax.plot(x, sigmoid(x), linestyle="-", label="sigmoid")
ax.plot(x, np.tanh(x), linestyle="--", label="tanh")
ax.plot(x, np.maximum(x, 0), linestyle="-.", label="ReLU")
ax.plot(x, np.logaddexp(0, x), linestyle=":", label="softplus")
ax.plot(x, softmax(x), linewidth=1, label="softmax over x")
ax.legend(loc="upper left")
fig.tight_layout()
```

## Swish

绘制 `swish(x) = x * sigmoid(beta * x)` 在不同 `beta` 下的图形。


```python
fig, ax = plt.subplots(figsize=(6, 4), dpi=100)
ax.spines["left"].set_position(("data", 0))
ax.spines["bottom"].set_position(("data", 0))
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.axis([-4.5, 4.5, -2.5, 4.5])

for beta, linestyle, color in [(0, "-", "g"), (0.5, "--", "c"), (1, "-.", "m"), (100, "-", "r")]:
    y = x * sigmoid(beta * x)
    ax.plot(x, y, linestyle=linestyle, color=color, label=rf"$\beta = {beta}$")

ax.legend()
fig.tight_layout()
```

读取 `resources/datasets/titanic.csv`，训练分组统计、条件筛选和存活率可视化。

## 数据读取

使用 `pd.read_csv()` 读取数据集。


```python
import matplotlib.pyplot as plt
import pandas as pd


titanic = pd.read_csv("../../resources/datasets/titanic.csv")
titanic.head()
```

## 统计任务

计算不同性别的平均年龄、整体平均存活率、小于 25 岁乘客存活率和不同性别存活率。


```python
print(titanic.groupby("Sex")["Age"].mean())
print(titanic["Survived"].mean())
print(titanic.loc[titanic["Age"] < 25, "Survived"].mean())
print(titanic.groupby("Sex")["Survived"].mean())
```

## 客舱等级存活率

按 `Pclass` 分组计算存活率，并绘制条形图。


```python
survival_by_class = titanic.groupby("Pclass")["Survived"].mean()
ax = survival_by_class.plot(kind="bar", figsize=(7, 4), rot=0)
ax.set_xlabel("Pclass")
ax.set_ylabel("Survival rate")
ax.set_ylim(0, 1)

for index, value in enumerate(survival_by_class):
    ax.text(index, value + 0.02, f"{value:.1%}", ha="center")

plt.tight_layout()
```

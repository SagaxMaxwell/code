## 绘制曲线

设置坐标轴穿过原点，使用 LaTeX 风格标签显示刻度和标注。


```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(-6, 6, 200)
point_x = 2 * np.pi / 3

fig, ax = plt.subplots(figsize=(6, 4), dpi=100, facecolor="w")
ax.spines["left"].set_position(("data", 0))
ax.spines["bottom"].set_position(("data", 0))
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.set_xlim(-3.5, 3.5)
ax.set_yticks([-1, 1])
ax.set_xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
ax.set_xticklabels([r"-$\pi$", r"-$\frac{\pi}{2}$", "0", r"$\frac{\pi}{2}$", r"$\pi$"])

ax.plot(x, np.sin(x), linewidth=2, color="r", label="sine")
ax.plot(x, np.cos(x), linewidth=2, color="b", label="cosine")
ax.plot([point_x, point_x], [np.sqrt(3) / 2, 0], linestyle="--", color="r")
ax.plot([point_x, point_x], [-0.5, 0], linestyle="--", color="b")
ax.scatter([point_x, point_x], [np.sqrt(3) / 2, -0.5], color="r")
ax.annotate(r"$\sin(\frac{2\pi}{3}) = \frac{\sqrt{3}}{2}$", xy=(point_x, np.sqrt(3) / 2), xytext=(point_x + 0.7, np.sqrt(3) / 2 + 0.2), arrowprops={"arrowstyle": "->", "color": "black"})
ax.annotate(r"$\cos(\frac{2\pi}{3}) = -\frac{1}{2}$", xy=(point_x, -0.5), xytext=(point_x - 2.2, -1), arrowprops={"arrowstyle": "->", "connectionstyle": "angle3, angleA=0, angleB=-120"})
ax.legend(loc="best")
fig.tight_layout()
```

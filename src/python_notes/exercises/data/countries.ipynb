# 国家数据练习

使用国家、人口、面积和首都数据，训练 DataFrame 创建、索引、筛选、列计算和基础可视化。

## 数据准备

创建 DataFrame，并将 `country` 列设置为索引。


```python
import matplotlib.pyplot as plt
import pandas as pd


data = {
    "country": ["Belgium", "France", "Germany", "Netherlands", "United Kingdom"],
    "population": [11.3, 64.3, 81.3, 16.9, 64.9],
    "area": [30510, 671308, 357050, 41526, 244820],
    "capital": ["Brussels", "Paris", "Berlin", "Amsterdam", "London"],
}

df = pd.DataFrame(data).set_index("country")
df
```

## 可视化

分别绘制各国人口和面积条形图。面积以 `10000 km^2` 为单位展示。


```python
plot_df = pd.DataFrame({"population_million": df["population"], "area_10k_km2": df["area"] / 10000})
ax = plot_df.plot(kind="bar", figsize=(9, 5), rot=30)
ax.set_xlabel("country")
ax.set_ylabel("value")
ax.set_title("Population and Area by Country")
plt.tight_layout()
```

## 筛选和列计算

完成面积筛选、人口密度计算、相对密度计算、首都修改和字符串筛选。


```python
print(df[df["area"] > 100000])

df["density"] = df["population"] * 1_000_000 / df["area"]
average_density = df["population"].sum() * 1_000_000 / df["area"].sum()
df["density_ratio"] = df["density"] / average_density
print(df.loc[df["density"] > 300, ["capital", "population"]])

df.loc["United Kingdom", "capital"] = "Cambridge"
print(df[df["capital"].str.len() > 7])
print(df[df["capital"].str.contains("am", case=False, na=False)])
```

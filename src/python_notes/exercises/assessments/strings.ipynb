## 字符串索引和切片

给定字符串 `Hello Python`，输出第一个字符、最后一个字符和子串 `Python`。


```python
print("Hello Python"[0])
print("Hello Python"[-1])
print("Hello Python"[6:])
```

## 奇数长度字符串

给定奇数长度字符串，输出中间字符、前半部分和后半部分。


```python
text = "123456789"
mid = len(text) // 2
print(text[mid])
print(text[:mid])
print(text[mid + 1:])
```

## 切片方向和反转

步长方向必须与起止位置一致，否则切片结果为空。


```python
text = "What is your name?"
print(text[::2])
print(text[2:8:-1])
print(text[::-1])
```

## 删除字符和统计字母

删除所有字符 `a`，并统计字符串中的元音和辅音数量。


```python
ab = "ababababababababab"
print(ab.replace("a", ""))

text = "Hello Python%^^%^&^%^&"
vowels = set("aeiou")
letters = {chr(i) for i in range(ord("a"), ord("z") + 1)}
consonants = letters - vowels
print(sum(char in vowels for char in text.lower()))
print(sum(char in consonants for char in text.lower()))
```

## 列表拼接和清理

把正整数列表转换为字符串，交叉连接两个列表，删除字符串列表中的空字符串，并按索引位置拼接两个列表。


```python
from itertools import product, zip_longest


print("".join(str(x) for x in [1, 2, 3, 4]))
print(["".join(pair) for pair in product(["Hello ", "take "], ["Dear", "Sir"])])
print([x for x in ["Mike", "", "Emma", "Kelly", "", "Brad"] if x])

list1 = ["M", "na", "i", "Ke"]
list2 = ["y", "me", "s", "lly"]
print([left + right for left, right in zip_longest(list1, list2, fillvalue="")])
```

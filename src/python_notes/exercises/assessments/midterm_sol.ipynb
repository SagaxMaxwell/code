5. 给定一个字符串，输出其中元音和辅音字母的个数。字符中可能包含除字母外的特殊字符。


````python
S = "Hello Python%^^%^&^%^&"
vowels = {"a", "e", "i", "o", "u"}
letters = {chr(i) for i in range(ord("a"), ord("z") + 1)}
consonants = letters - vowels

vowel_count = sum(char in vowels for char in S.lower())
consonant_count = sum(char in consonants for char in S.lower())

print(f"元音：{vowel_count}")
print(f"辅音：{consonant_count}")
````

```python
["Hello Dear", "Hello Sir", "take Dear", "take Sir"]
```

````python
from itertools import product

lst1 = ["Hello ", "take "]
lst2 = ["Dear", "Sir"]
for combination in product(lst1, lst2):
    print("".join(combination))
````

13. 编写 `factorial(num)` 函数，返回整数 `num` 的阶乘，并用测试数据调用该函数。


````python
def factorial(num):
    """计算非负整数的阶乘。

    Args:
        num: 待计算阶乘的非负整数。

    Returns:
        `num` 的阶乘结果。
    """
    result = 1
    for number in range(2, num + 1):
        result *= number
    return result


print(factorial(5))
````

14. 编写一个函数 `sum_range(x, y)`，输入 `x` 和 `y` 两个数字，并返回从 `x` 到 `y` 之间的所有数字之和。


````python
def sum_range(x, y):
    """计算闭区间内所有整数之和。

    Args:
        x: 区间端点之一。
        y: 区间端点之一。

    Returns:
        从较小端点到较大端点之间所有整数的和。
    """
    start, end = sorted((x, y))
    return sum(range(start, end + 1))


print(sum_range(1, 5))
````

15. 定义一个 `Python` 函数 `box_volume(length, height, depth)`，输入三个参数：长度、高度和深度，返回长方体的体积。


````python
def box_volume(length, height, depth):
    """计算长方体体积。

    Args:
        length: 长方体长度。
        height: 长方体高度。
        depth: 长方体深度。

    Returns:
        长方体体积。
    """
    return length * height * depth


print(box_volume(2, 3, 4))
````

16. 定义一个 `Python` 函数 `strictly_increasing(n1, n2, n3, n4)`。当四个参数满足 `n1 < n2 < n3 < n4` 时返回 `True`，否则返回 `False`。


````python
def strictly_increasing(n1, n2, n3, n4):
    """判断四个数是否严格递增。

    Args:
        n1: 第一个数。
        n2: 第二个数。
        n3: 第三个数。
        n4: 第四个数。

    Returns:
        如果满足 `n1 < n2 < n3 < n4`，返回 `True`；否则返回 `False`。
    """
    return n1 < n2 < n3 < n4


print(strictly_increasing(3, 4, 2, 1))
print(strictly_increasing(-100, 33, 66, 100))
````

17. 编写函数 `get_word_frequencies()`，读取字符串列表中的每个单词，并返回一个计数字典。该字典将单词映射到出现次数。要求忽略 `common_words` 中的常见词，并且大小写不敏感，例如 `"Science"` 和 `"science"` 视为同一个单词。

示例输入：

```python
string_list = [
    "Your demure expression",
    "the unfailing grace with which you meet",
    "the small misfortunes which we meet each day",
    "your ready smile",
    "intelligent gaze",
]
common_words = ["the", "a", "an", "that", "this"]
```

示例输出：

```python
word_counts = {
    "your": 2,
    "demure": 1,
    "expression": 1,
    "unfailing": 1,
    "grace": 1,
    "with": 1,
    "which": 2,
    "you": 1,
    "meet": 2,
    "small": 1,
    "misfortunes": 1,
    "we": 1,
    "each": 1,
    "day": 1,
    "ready": 1,
    "smile": 1,
    "intelligent": 1,
    "gaze": 1,
}
```


````python
from collections import Counter


def get_word_frequencies(string_list, common_words):
    """统计单词出现次数。

    Args:
        string_list: 待统计的字符串列表。
        common_words: 需要忽略的常见词列表。

    Returns:
        单词到出现次数的映射，统计过程不区分大小写。
    """
    stop_words = {word.lower() for word in common_words}
    words = " ".join(string_list).lower().split()
    return dict(Counter(word for word in words if word not in stop_words))


string_list = [
    "Your demure expression",
    "the unfailing grace with which you meet",
    "the small misfortunes which we meet each day",
    "your ready smile",
    "intelligent gaze",
]
common_words = ["the", "a", "an", "that", "this"]
print(get_word_frequencies(string_list, common_words))
````

18. 编写函数 `get_n_most_frequent()`，输入整数 `n` 和单词计数字典 `word_counts`，返回计数最高的 `n` 个单词。返回值是元组列表，每个元组包含单词和出现次数，并按出现次数降序排列。

示例输入：

```python
word_counts = {"science": 2, "fun": 1, "karel": 1, "computer": 3, "gates": 1}
get_n_most_frequent(word_counts, 2)
```

期望输出：

```python
[("computer", 3), ("science", 2)]
```


````python
def get_n_most_frequent(word_counts, n):
    """获取出现次数最高的单词。

    Args:
        word_counts: 单词计数字典。
        n: 需要返回的单词数量。

    Returns:
        按出现次数降序排列的 `(word, count)` 元组列表。
    """
    return sorted(word_counts.items(), key=lambda item: item[1], reverse=True)[:n]


word_counts = {"science": 2, "fun": 1, "karel": 1, "computer": 3, "gates": 1}
print(get_n_most_frequent(word_counts, 2))
````

19. `Pig Latin` 是一种单词改写游戏。规则如下：

- 如果单词以元音开头，则在单词末尾添加 `"yay"`。
- 如果单词以辅音开头，则把从开头到第一个元音之前的辅音字母移动到单词末尾，并添加 `"ay"`。

示例：

| 原单词 | 转换结果 |
| :--- | :--- |
| `dog` | `ogday` |
| `scratch` | `atchscray` |
| `is` | `isyay` |
| `apple` | `appleyay` |

请编写程序，反复提示用户输入英文单词，并输出转换后的 `Pig Latin` 单词。用户输入句号 `.` 时，程序终止。


````python
def pig_latin(word):
    """将英文单词转换为 `Pig Latin`。

    Args:
        word: 待转换的英文单词。

    Returns:
        转换后的 `Pig Latin` 单词。
    """
    vowels = set("aeiou")
    if word[0].lower() in vowels:
        return word + "yay"

    for index, char in enumerate(word):
        if char.lower() in vowels:
            return word[index:] + word[:index] + "ay"
    return word + "ay"


print([pig_latin(word) for word in ["dog", "scratch", "is", "apple"]])
````

20. 编写 `Python` 程序实现类。
- 创建一个 `Vehicle` 类，该类具有实例属性 `name`，`capacity` 和 `mileage`；类属性 `color`；类方法 `fare()` 计算费用。车辆费用等于容量乘以 100，即返回 `capacity * 100`。
- 创建一个 `Bus` 类，继承 `Vehicle` 类。要求重写父类的 `fare()` 方法，使得费用是原来的基础上加收 `10%`。


````python
class Vehicle:
    """表示普通车辆。

    Attributes:
        color: 车辆颜色。
    """

    color = "red"

    def __init__(self, name, capacity, mileage):
        """初始化车辆对象。

        Args:
            name: 车辆名称。
            capacity: 载客容量。
            mileage: 行驶里程。
        """
        self.name = name
        self.capacity = capacity
        self.mileage = mileage

    def fare(self):
        """计算车辆费用。

        Returns:
            按载客容量计算的基础费用。
        """
        return self.capacity * 100


class Bus(Vehicle):
    """表示公交车。"""

    def fare(self):
        """计算公交车费用。

        Returns:
            在基础费用上加收 `10%` 后的费用。
        """
        return super().fare() * 1.1


print(Vehicle("car", 4, 10000).fare())
print(Bus("bus", 40, 20000).fare())
````

21. 给定一个二进制字符串，编写函数 `find_max_length(binary_string)`，返回其中最长连续 `1` 子串的长度。


````python
line = "0111011010100111110000010111111"


def find_max_length(binary_string):
    """查找最长连续 `1` 子串长度。

    Args:
        binary_string: 仅包含 `0` 和 `1` 的字符串。

    Returns:
        最长连续 `1` 子串的长度。
    """
    return max(len(part) for part in binary_string.split("0"))


print(find_max_length(line))
````

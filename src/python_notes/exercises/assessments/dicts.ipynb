## 字典最小值

从字典中获取值最小的键。


```python
sample_dict = {"Physics": 82, "Math": 65, "history": 75}
print(min(sample_dict, key=sample_dict.get))
```

## 字典筛选

给定键集合，分别取出指定键和删除指定键。


```python
sample_dict = {"name": "Kelly", "age": 25, "salary": 8000, "city": "New york"}
keys = {"name", "salary"}
print({key: value for key, value in sample_dict.items() if key in keys})
print({key: value for key, value in sample_dict.items() if key not in keys})
```

## 条件判断

输入三个数字，查找最小值。Notebook 中使用固定示例避免交互输入。


```python
numbers = [3, 1, 2]
print(min(numbers))
```

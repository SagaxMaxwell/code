## Vehicle 和 Bus

创建 `Vehicle` 类，包含 `name`、`capacity`、`mileage` 和类属性 `color`；创建继承自 `Vehicle` 的 `Bus` 类，并重写费用计算方法。


```python
class Vehicle:
    color = "red"

    def __init__(self, name, capacity, mileage):
        self.name = name
        self.capacity = capacity
        self.mileage = mileage

    def fare(self):
        return self.capacity * 100


class Bus(Vehicle):
    def fare(self):
        return super().fare() * 1.1


print(Vehicle("car", 4, 10000).fare())
print(Bus("bus", 40, 20000).fare())
```

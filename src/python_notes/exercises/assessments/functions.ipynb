## 数值函数

编写函数计算阶乘、闭区间求和、长方体体积和严格递增判断。


```python
def factorial(num):
    result = 1
    for number in range(2, num + 1):
        result *= number
    return result


def sum_range(x, y):
    start, end = sorted((x, y))
    return sum(range(start, end + 1))


def box_volume(length, height, depth):
    return length * height * depth


def strictly_increasing(n1, n2, n3, n4):
    return n1 < n2 < n3 < n4


print(factorial(5))
print(sum_range(1, 5))
print(box_volume(2, 3, 4))
print(strictly_increasing(-100, 33, 66, 100))
```

## 词频统计

忽略常见词并且大小写不敏感，统计单词出现次数，再返回出现次数最高的 `n` 个单词。


```python
from collections import Counter


def get_word_frequencies(string_list, common_words):
    stop_words = {word.lower() for word in common_words}
    words = " ".join(string_list).lower().split()
    return dict(Counter(word for word in words if word not in stop_words))


def get_n_most_frequent(word_counts, n):
    return sorted(word_counts.items(), key=lambda item: item[1], reverse=True)[:n]


strings = ["Your demure expression", "the small misfortunes which we meet each day", "your ready smile"]
counts = get_word_frequencies(strings, ["the", "a", "an", "that", "this"])
print(counts)
print(get_n_most_frequent(counts, 2))
```

## Pig Latin

如果单词以元音开头，在末尾添加 `yay`；如果以辅音开头，把开头到第一个元音之前的辅音移动到末尾并添加 `ay`。


```python
def pig_latin(word):
    vowels = set("aeiou")
    if word[0].lower() in vowels:
        return word + "yay"
    for index, char in enumerate(word):
        if char.lower() in vowels:
            return word[index:] + word[:index] + "ay"
    return word + "ay"


print([pig_latin(word) for word in ["dog", "scratch", "is", "apple"]])
```

## 连续 1

给定二进制字符串，返回最长连续 `1` 子串长度。


```python
def find_max_length(binary_string):
    return max(len(part) for part in binary_string.split("0"))


print(find_max_length("0111011010100111110000010111111"))
```
